```r
estimate_cell_cycle = function(cds, g1s_markers, g2m_markers){
  cds_g1s = cds[rowData(cds)$gene_short_name %in% g1s_markers,] 
  aggregate_g1s_expression = counts(cds_g1s)
  aggregate_g1s_expression = t(t(aggregate_g1s_expression) / colData(cds_g1s)$Size_Factor)
  aggregate_g1s_expression = Matrix::colSums(aggregate_g1s_expression)
  
  cds_g2m = cds[rowData(cds)$gene_short_name %in% g2m_markers,] 
  aggregate_g2m_expression = counts(cds_g2m)
  aggregate_g2m_expression = t(t(aggregate_g2m_expression) / colData(cds_g2m)$Size_Factor)
  aggregate_g2m_expression = Matrix::colSums(aggregate_g2m_expression)
  #pData(cds)$cell_cycle_score = aggregate_g1s_expression
  cds$g1s_score = log(aggregate_g1s_expression+1)
  cds$g2m_score = log(aggregate_g2m_expression+1)
  cds$proliferation_index = log(aggregate_g1s_expression + aggregate_g2m_expression + 1)
  return(cds)
}

aggregate_gene_score = function(cds, gene_list=c("CCNB1", "CCNB2","CDK1"), name = "gene_list_score"){
  #gene_list = stringr::str_to_title(gene_list)
  cds_gene_list_norm = cds[rowData(cds)$gene_short_name %in% gene_list,] 
  aggregate_marker_expression = counts(cds_gene_list_norm)
  aggregate_marker_expression = t(t(aggregate_marker_expression) / colData(cds_gene_list_norm)$Size_Factor)
  aggregate_marker_expression = Matrix::colSums(aggregate_marker_expression)
  aggregate_marker_expression = log(aggregate_marker_expression + 1)
  #aggregate_marker_expression = t(scale(t(aggregate_marker_expression)))
  #aggregate_marker_expression = aggregate_marker_expression[rowSums(is.na(aggregate_marker_expression)) == 0,]
  #aggregate_marker_expression = colMeans(aggregate_marker_expression)
  colData(cds)[[name]] = aggregate_marker_expression
  return(cds)
}
```

```r
all_correlations_df = metadata %>%
    filter(!trajectory_label %in%traj_to_exclude) %>%
    filter(!is.na(Pseudotime_avg)) %>%
    filter(
        Phenotype %in% c("Area", "Ratio"),
        Marker %in% c("Nucleolin", "SC35")
    ) %>%
    group_by(Embryo, trajectory_label, Cluster_sub_2, Marker, Phenotype) %>%
    summarise(
        med_score = median(avg_score, na.rm = TRUE),
        med_pseudotime = median(Pseudotime_avg, na.rm = TRUE),
        .groups = "drop"
    ) %>%
    
    mutate(Embryo = as.factor(Embryo)) %>%
    group_by(trajectory_label, Marker, Phenotype) %>%
    add_tally() %>%

    filter(n_distinct(Embryo) > 1) %>%
    
    mutate(
        ranked_pseudotime = rank(med_pseudotime, ties.method = "average"),
        ranked_score = rank(med_score, ties.method = "average")
    ) %>%

    summarise(
        # 1. Fit the Pearson model safely (Store it in a temporary list column)
        pearson_model = list(
            tryCatch(
                rmcorr(participant = Embryo, measure1 = med_pseudotime, measure2 = med_score, dataset = pick(everything())),
                error = function(e) list(r = NA_real_, p = NA_real_) # Return NAs if it fails
            )
        ),

        # 2. Fit the Spearman model safely
        spearman_model = list(
            tryCatch(
                rmcorr(participant = Embryo, measure1 = ranked_pseudotime, measure2 = ranked_score, dataset = pick(everything())),
                error = function(e) list(r = NA_real_, p = NA_real_)
            )
        ),

        # 3. Extract the R and P values directly from the stored models
        pearson_rmc     = pearson_model[[1]]$r,
        pearson_pvalue  = pearson_model[[1]]$p,

        spearman_rmc    = spearman_model[[1]]$r,
        spearman_pvalue = spearman_model[[1]]$p,

        .groups = "drop"
    ) %>%
    # 4. Remove the temporary model columns to keep your dataframe clean
    select(-pearson_model, -spearman_model) %>%
    
    # 5. Create the Final Labels
    mutate(
        suffix = ifelse(Phenotype == "Area", "abs", "rel"),
        marker = paste0(Marker, "-", suffix)
    ) %>%
    select(-suffix)
```

```r
color = colorRampPalette(c("navy", "white", "red"))(101)
breaks = seq(-1, 1, length.out = 101)

options(repr.plot.width=4.5, repr.plot.height=3.7, repr.plot.res=300)

# 1. Create the Correlation Matrix (Colors)
cor_matrix <- all_correlations_df %>%
    select(trajectory_label, spearman_rmc, marker) %>%
    tidyr::spread(marker, spearman_rmc) %>%
    # RECODE FIRST (while the underscores still exist)
    mutate(trajectory_label = recode(trajectory_label, 
                "Definitive_erythroid" = "Definitive erythroids",
                "GABAergic" = "GABAergic neurons",
                "Glutamatergic" = "Glutamatergic neurons",
                "Fibroblasts_Fap+" = "Fibroblasts (Fap+)")) %>%
    # THEN GSUB to clean up whatever is left
    mutate(trajectory_label = gsub('_', ' ', trajectory_label)) %>%
    column_to_rownames("trajectory_label") %>%
    as.matrix()

# 2. Create the P-value Matrix (Significance)
pval_matrix <- all_correlations_df %>%
    select(trajectory_label, spearman_pvalue, marker) %>%
    tidyr::spread(marker, spearman_pvalue) %>%
    # MUST MATCH COR_MATRIX EXACTLY
    mutate(trajectory_label = recode(trajectory_label, 
                "Definitive_erythroid" = "Definitive erythroids",
                "GABAergic" = "GABAergic neurons",
                "Glutamatergic" = "Glutamatergic neurons",
                "Fibroblasts_Fap+" = "Fibroblasts (Fap+)")) %>%
    mutate(trajectory_label = gsub('_', ' ', trajectory_label)) %>%
    column_to_rownames("trajectory_label") %>%
    as.matrix()

# 3. Convert P-values into Asterisks
sig_matrix <- ifelse(pval_matrix < 0.001, "***", 
              ifelse(pval_matrix < 0.01, "**", 
              ifelse(pval_matrix < 0.05, "*", "")))

# Safety net
sig_matrix[is.na(sig_matrix)] <- ""

# 4. Plot
ComplexHeatmap::pheatmap(
    cor_matrix, 
    cluster_cols = F, 
    cluster_rows = T, 
    
    color = color, 
    breaks = breaks, 
    
    display_numbers = sig_matrix,    
    fontsize_number = 12,          
    number_color = "white",      
    
    na_col = "grey80",
    
    border_color = "white",        
    cellwidth = 22,                
    cellheight = 20,               
    
    fontsize_row = 9, 
    fontsize_col = 9,
    #angle_col = 45,
    
    treeheight_row = 20,
    
    show_colnames = FALSE,
    heatmap_legend_param = list(legend_height = unit(4.5, "cm")) #direction = "horizontal")
)
```

```r
options(repr.plot.width=3.5, repr.plot.height=3, repr.plot.res=300)

# 1. BASE DATA
base_data <- metadata_erythroids %>%
    group_by(Experiment, trajectory_refined, Embryo, Cluster_sub) %>%
    summarise(
        med_prolif = median(proliferation_index, na.rm = TRUE),
        med_Pseudotime_all = median(Pseudotime_avg, na.rm = TRUE),
        med_score_filtered = median(avg_score[Marker == "Nucleolin" & Phenotype == "Ratio"], na.rm = TRUE),
        med_Pseudotime_filtered = median(Pseudotime_avg[Marker == "Nucleolin" & Phenotype == "Ratio"], na.rm = TRUE),
        .groups = "drop"
    ) %>%
    filter(!is.na(med_score_filtered) & !is.na(med_prolif)) %>%
    mutate(
        norm_prolif = normalize(med_prolif),
        norm_score = normalize(med_score_filtered)
    )

# 2. COLLAPSED POINTS (Min-Max normalization applied to the collapsed points)
points_score <- base_data %>%
    group_by(med_Pseudotime_filtered) %>%
    summarise(collapsed_score = median(norm_score, na.rm = TRUE), .groups = "drop") %>%
    mutate(collapsed_score_norm = (collapsed_score - min(collapsed_score, na.rm=T)) / 
                                  (max(collapsed_score, na.rm=T) - min(collapsed_score, na.rm=T)))

points_prolif <- base_data %>%
    group_by(med_Pseudotime_all) %>%
    summarise(collapsed_prolif = median(norm_prolif, na.rm = TRUE), .groups = "drop") %>%
    mutate(collapsed_prolif_norm = (collapsed_prolif - min(collapsed_prolif, na.rm=T)) / 
                                   (max(collapsed_prolif, na.rm=T) - min(collapsed_prolif, na.rm=T)))


# 3. MANUALLY COMPUTE AND NORMALIZE LOESS CURVES WITH STANDARD ERROR (se = TRUE)
clean_score_data <- as.data.frame(base_data) %>% 
    filter(!is.na(norm_score) & !is.na(med_Pseudotime_filtered))
clean_prolif_data <- as.data.frame(base_data) %>% 
    filter(!is.na(norm_prolif) & !is.na(med_Pseudotime_all))

fit_score <- loess(norm_score ~ med_Pseudotime_filtered, data = clean_score_data, span = 0.9)
fit_prolif <- loess(norm_prolif ~ med_Pseudotime_all, data = clean_prolif_data, span = 0.9)

loess_score_data <- data.frame(
    med_Pseudotime_filtered = seq(min(clean_score_data$med_Pseudotime_filtered), 
                                  max(clean_score_data$med_Pseudotime_filtered), length.out = 100)
)
loess_prolif_data <- data.frame(
    med_Pseudotime_all = seq(min(clean_prolif_data$med_Pseudotime_all), 
                             max(clean_prolif_data$med_Pseudotime_all), length.out = 100)
)

# Predict the Y values AND extract standard errors
pred_score <- predict(fit_score, newdata = loess_score_data, se = TRUE)
pred_prolif <- predict(fit_prolif, newdata = loess_prolif_data, se = TRUE)

# Calculate 95% Confidence Interval Bounds (1.96 * SE)
loess_score_data <- loess_score_data %>%
    mutate(
        y = pred_score$fit,
        y_upper = pred_score$fit + (1.96 * pred_score$se.fit),
        y_lower = pred_score$fit - (1.96 * pred_score$se.fit)
    )

loess_prolif_data <- loess_prolif_data %>%
    mutate(
        y = pred_prolif$fit,
        y_upper = pred_prolif$fit + (1.96 * pred_prolif$se.fit),
        y_lower = pred_prolif$fit - (1.96 * pred_prolif$se.fit)
    )

# Apply Min-Max Normalization to the line AND the ribbons simultaneously
min_score_line <- min(loess_score_data$y, na.rm = TRUE)
max_score_line <- max(loess_score_data$y, na.rm = TRUE)

loess_score_data <- loess_score_data %>%
    mutate(
        y_norm = (y - min_score_line) / (max_score_line - min_score_line),
        y_upper_norm = (y_upper - min_score_line) / (max_score_line - min_score_line),
        y_lower_norm = (y_lower - min_score_line) / (max_score_line - min_score_line)
    )

min_prolif_line <- min(loess_prolif_data$y, na.rm = TRUE)
max_prolif_line <- max(loess_prolif_data$y, na.rm = TRUE)

loess_prolif_data <- loess_prolif_data %>%
    mutate(
        y_norm = (y - min_prolif_line) / (max_prolif_line - min_prolif_line),
        y_upper_norm = (y_upper - min_prolif_line) / (max_prolif_line - min_prolif_line),
        y_lower_norm = (y_lower - min_prolif_line) / (max_prolif_line - min_prolif_line)
    )


# 4. PLOT (Ribbons + Lines + Collapsed Normalized Points)
ggplot() + 
    
    # --- CONFIDENCE INTERVAL RIBBONS (Drawn first so lines/points sit on top) ---
    geom_ribbon(data = loess_score_data, 
                aes(x = med_Pseudotime_filtered, ymin = y_lower_norm, ymax = y_upper_norm), 
                fill = "grey70", alpha = 0.4) + 
                
    geom_ribbon(data = loess_prolif_data, 
                aes(x = med_Pseudotime_all, ymin = y_lower_norm, ymax = y_upper_norm), 
                fill = "#fb9a99", alpha = 0.25) + 
                
    # --- LOESS LINES ---
    geom_line(data = loess_score_data, 
              aes(x = med_Pseudotime_filtered, y = y_norm), 
              linewidth = 1.2, color = "black") + 
              
    geom_line(data = loess_prolif_data, 
              aes(x = med_Pseudotime_all, y = y_norm), 
              linewidth = 1.2, color = "#fb9a99") + 
    
    # --- POINTS (Plotting the 0-to-1 normalized collapsed points) ---
    geom_point(data = points_score, 
               aes(x = med_Pseudotime_filtered, y = collapsed_score_norm), 
               size = 1.5, shape = 3, color = "black") + 
               
    geom_point(data = points_prolif, 
               aes(x = med_Pseudotime_all, y = collapsed_prolif_norm), 
               size = 1.5, shape = 17, color = "black") +
    
    labs(x = "Pseudotime", y = "Normalized score") + 
    monocle3:::monocle_theme_opts() + 
    theme(legend.position = "none",
          axis.text.x = element_text(color="black", size = 11), 
          axis.text.y = element_text(color = "black", size = 11),
          axis.ticks = element_line(color = "black"),
          aspect.ratio = 1)
```

```r
options(repr.plot.width=4, repr.plot.height=3.5, repr.plot.res=300)

# --- 1. DATA AGGREGATION (THE JOIN METHOD) ---
# We calculate each stream separately to ensure NAs don't drop rows prematurely

# Proliferation (Base)
prolif_base <- metadata_erythroids %>%
    group_by(Experiment, trajectory_refined, Embryo, Cluster_sub) %>%
    summarise(med_prolif = median(proliferation_index, na.rm = TRUE),
              pt_prolif  = median(Pseudotime_avg, na.rm = TRUE), .groups = "drop")

# Nucleolin
nuc_base <- metadata_erythroids %>%
    filter(Marker == "Nucleolin" & Phenotype == "Ratio") %>%
    group_by(Experiment, trajectory_refined, Embryo, Cluster_sub) %>%
    summarise(med_nuc = median(avg_score, na.rm = TRUE),
              pt_nuc  = median(Pseudotime_avg, na.rm = TRUE), .groups = "drop")

# SC35
sc35_base <- metadata_erythroids %>%
    filter(Marker == "SC35" & Phenotype == "Ratio") %>%
    group_by(Experiment, trajectory_refined, Embryo, Cluster_sub) %>%
    summarise(med_sc35 = median(avg_score, na.rm = TRUE),
              pt_sc35  = median(Pseudotime_avg, na.rm = TRUE), .groups = "drop")

# 1. Merge the data
base_data <- prolif_base %>%
    full_join(nuc_base,  by = c("Experiment", "trajectory_refined", "Embryo", "Cluster_sub")) %>%
    full_join(sc35_base, by = c("Experiment", "trajectory_refined", "Embryo", "Cluster_sub"))

# 2. Normalize with NA safety
# This ensures that NAs created by the join don't break the 0-1 scaling
base_data <- base_data %>%
    mutate(
        norm_prolif = (med_prolif - min(med_prolif, na.rm = TRUE)) / 
                      (max(med_prolif, na.rm = TRUE) - min(med_prolif, na.rm = TRUE)),
        
        norm_nuc    = (med_nuc - min(med_nuc, na.rm = TRUE)) / 
                      (max(med_nuc, na.rm = TRUE) - min(med_nuc, na.rm = TRUE)),
        
        norm_sc35   = (med_sc35 - min(med_sc35, na.rm = TRUE)) / 
                      (max(med_sc35, na.rm = TRUE) - min(med_sc35, na.rm = TRUE))
    )

# --- 2. MEDIAN-ANCHORED SCALING ---
# Helper to get the Min/Max of the collapsed cluster medians
get_anchors <- function(df, pt_col, val_col) {
    temp <- df %>% filter(!is.na(!!sym(val_col))) %>% 
            group_by(!!sym(pt_col)) %>% 
            summarise(v = median(!!sym(val_col), na.rm = TRUE))
    return(c(min(temp$v, na.rm=T), max(temp$v, na.rm=T)))
}

anc_pro  <- get_anchors(base_data, "pt_prolif", "norm_prolif")
anc_nuc  <- get_anchors(base_data, "pt_nuc", "norm_nuc")
anc_sc35 <- get_anchors(base_data, "pt_sc35", "norm_sc35")

# Apply scaling to the raw points based on those anchors
base_data <- base_data %>%
    mutate(
        scaled_pro  = (norm_prolif - anc_pro[1]) / (anc_pro[2] - anc_pro[1]),
        scaled_nuc  = (norm_nuc - anc_nuc[1]) / (anc_nuc[2] - anc_nuc[1]),
        scaled_sc35 = (norm_sc35 - anc_sc35[1]) / (anc_sc35[2] - anc_sc35[1])
    )

# --- 3. MANUALLY COMPUTE NORMALISED LOESS CURVES ---
# Helper function to fit, predict, and normalize 0-1
get_loess_df <- function(df, pt_col, val_col) {
    clean <- df %>% filter(!is.na(!!sym(val_col)) & !is.na(!!sym(pt_col)))
    fit   <- loess(as.formula(paste(val_col, "~", pt_col)), data = clean, span = 0.9)
    res   <- data.frame(pt = seq(min(clean[[pt_col]]), max(clean[[pt_col]]), length.out = 100))
    colnames(res) <- pt_col
    
    pred  <- predict(fit, newdata = res, se = TRUE)
    res$y <- pred$fit
    res$y_u <- pred$fit + (1.96 * pred$se.fit)
    res$y_l <- pred$fit - (1.96 * pred$se.fit)
    
    # Normalize the curve itself 0 to 1
    mn <- min(res$y, na.rm=T); mx <- max(res$y, na.rm=T)
    res$y_norm   <- (res$y - mn) / (mx - mn)
    res$y_u_norm <- (res$y_u - mn) / (mx - mn)
    res$y_l_norm <- (res$y_l - mn) / (mx - mn)
    return(res)
}

loess_pro  <- get_loess_df(base_data, "pt_prolif", "norm_prolif")
loess_nuc  <- get_loess_df(base_data, "pt_nuc", "norm_nuc")
loess_sc35 <- get_loess_df(base_data, "pt_sc35", "norm_sc35")

# --- 4. PLOT ALL THREE TRENDS ---
ggplot() + 
    # Ribbons (Standard Errors)
    geom_ribbon(data = loess_pro, aes(x = pt_prolif, ymin = y_l_norm, ymax = y_u_norm), fill = "#fb9a99", alpha = 0.2) + 
    geom_ribbon(data = loess_nuc, aes(x = pt_nuc, ymin = y_l_norm, ymax = y_u_norm), fill = "grey70", alpha = 0.3) + 
    geom_ribbon(data = loess_sc35, aes(x = pt_sc35, ymin = y_l_norm, ymax = y_u_norm), fill = "#1f78b4", alpha = 0.15) + 
    
    # Loess Lines
    geom_line(data = loess_pro, aes(x = pt_prolif, y = y_norm), color = "#fb9a99", linewidth = 1.2) + 
    geom_line(data = loess_nuc, aes(x = pt_nuc, y = y_norm), color = "black", linewidth = 1.2) + 
    geom_line(data = loess_sc35, aes(x = pt_sc35, y = y_norm), color = "#1f78b4", linewidth = 1.2) + 
    
    # Points (Scaled by median anchors)
    geom_point(data = base_data, aes(x = pt_prolif, y = scaled_pro), shape = 17, size = 1.2, color = "black") + 
    geom_point(data = base_data, aes(x = pt_nuc, y = scaled_nuc), shape = 3, size = 1.2, color = "black") + 
    geom_point(data = base_data, aes(x = pt_sc35, y = scaled_sc35), shape = 18, size = 1.5, color = "black") + 
    
    # Styling
    labs(x = "Pseudotime", y = "Median-Anchored Normalized Value") + 
    monocle3:::monocle_theme_opts() + 
    theme(legend.position = "none",
          axis.text = element_text(color = "black", size = 10),
          axis.ticks = element_line(color = "black"))
```

```r
options(repr.plot.width=3.5, repr.plot.height=4, repr.plot.res=300)

# 1. DATA EXTRACTION (Separate streams to handle NAs)
pro_df <- metadata_erythroids %>%
    group_by(Experiment, trajectory_refined, Embryo, Cluster_sub) %>%
    summarise(med_prolif = median(proliferation_index, na.rm = TRUE),
              pt_prolif  = median(Pseudotime_avg, na.rm = TRUE), .groups = "drop")

nuc_df <- metadata_erythroids %>%
    filter(Marker == "Nucleolin" & Phenotype == "Area") %>%
    group_by(Experiment, trajectory_refined, Embryo, Cluster_sub) %>%
    summarise(med_nuc = median(avg_score, na.rm = TRUE),
              pt_nuc  = median(Pseudotime_avg, na.rm = TRUE), .groups = "drop")

sc35_df <- metadata_erythroids %>%
    filter(Marker == "SC35" & Phenotype == "Area") %>%
    group_by(Experiment, trajectory_refined, Embryo, Cluster_sub) %>%
    summarise(med_sc35 = median(avg_score, na.rm = TRUE),
              pt_sc35  = median(Pseudotime_avg, na.rm = TRUE), .groups = "drop")

# 2. JOIN & MEDIAN-ANCHORED SCALING
base_data <- pro_df %>%
    full_join(nuc_df,  by = c("Experiment", "trajectory_refined", "Embryo", "Cluster_sub")) %>%
    full_join(sc35_base, by = c("Experiment", "trajectory_refined", "Embryo", "Cluster_sub")) %>%
    mutate(
        norm_pro = (med_prolif - min(med_prolif, na.rm=T)) / (max(med_prolif, na.rm=T) - min(med_prolif, na.rm=T)),
        norm_nuc = (med_nuc - min(med_nuc, na.rm=T)) / (max(med_nuc, na.rm=T) - min(med_nuc, na.rm=T)),
        norm_sc35 = (med_sc35 - min(med_sc35, na.rm=T)) / (max(med_sc35, na.rm=T) - min(med_sc35, na.rm=T))
    )

get_anchors <- function(df, pt, val) {
    tmp <- df %>% filter(!is.na(!!sym(val))) %>% group_by(!!sym(pt)) %>% summarise(v=median(!!sym(val)))
    return(c(min(tmp$v, na.rm=T), max(tmp$v, na.rm=T)))
}

anc_pro <- get_anchors(base_data, "pt_prolif", "norm_pro")
anc_nuc <- get_anchors(base_data, "pt_nuc", "norm_nuc")
anc_sc35 <- get_anchors(base_data, "pt_sc35", "norm_sc35")

base_data <- base_data %>%
    mutate(
        scaled_pro  = (norm_pro - anc_pro[1]) / (anc_pro[2] - anc_pro[1]),
        scaled_nuc  = (norm_nuc - anc_nuc[1]) / (anc_nuc[2] - anc_nuc[1]),
        scaled_sc35 = (norm_sc35 - anc_sc35[1]) / (anc_sc35[2] - anc_sc35[1])
    )

# 3. LOESS CURVES (Normalized 0-1)
get_loess_final <- function(df, pt, val) {
    clean <- df[!is.na(df[[pt]]) & !is.na(df[[val]]), ]
    fit <- loess(as.formula(paste(val, "~", pt)), data = clean, span = 0.9)
    res <- data.frame(x = seq(min(clean[[pt]]), max(clean[[pt]]), length.out = 100))
    colnames(res) <- pt
    pred <- predict(fit, newdata = res, se = TRUE)
    mn <- min(pred$fit); mx <- max(pred$fit)
    res$y_n   <- (pred$fit - mn) / (mx - mn)
    res$y_u_n <- ((pred$fit + 1.96*pred$se.fit) - mn) / (mx - mn)
    res$y_l_n <- ((pred$fit - 1.96*pred$se.fit) - mn) / (mx - mn)
    return(res)
}

l_pro  <- get_loess_final(as.data.frame(base_data), "pt_prolif", "norm_pro")
l_nuc  <- get_loess_final(as.data.frame(base_data), "pt_nuc", "norm_nuc")
l_sc35 <- get_loess_final(as.data.frame(base_data), "pt_sc35", "norm_sc35")

# 4. PLOT (Layered for maximum clarity)
ggplot() +
    # Layer 1: Ribbons (Mapped to 'fill' for the legend)
    geom_ribbon(data = l_pro,  aes(x=pt_prolif, ymin=y_l_n, ymax=y_u_n, fill="Proliferation"), alpha=0.15) +
    geom_ribbon(data = l_nuc,  aes(x=pt_nuc,    ymin=y_l_n, ymax=y_u_n, fill="Nucleolin"),     alpha=0.1) +
    geom_ribbon(data = l_sc35, aes(x=pt_sc35,   ymin=y_l_n, ymax=y_u_n, fill="SC35"),          alpha=0.1) +
    
    ## Layer 2: Points (Mapped to 'color' to match the lines)
    #geom_point(data = base_data, aes(x=pt_prolif, y=scaled_pro,  color="Proliferation"), shape=17, size=1, alpha=0.4) +
    #geom_point(data = base_data, aes(x=pt_nuc,    y=scaled_nuc,  color="Nucleolin"),     shape=3,  size=1, alpha=0.3) +
    #geom_point(data = base_data, aes(x=pt_sc35,   y=scaled_sc35, color="SC35"),          shape=18, size=1, alpha=0.3) +
    
    # Layer 3: Lines (Mapped to 'color' for the legend)
    geom_line(data = l_pro,  aes(x=pt_prolif, y=y_n, color="Proliferation"), linewidth=1.3) +
    geom_line(data = l_nuc,  aes(x=pt_nuc,    y=y_n, color="Nucleolin"),     linewidth=1.3) +
    geom_line(data = l_sc35, aes(x=pt_sc35,   y=y_n, color="SC35"),          linewidth=1.3) +
    
    
    # Define the manual color palette
    scale_color_manual(name = "Marker", 
                       values = c("Proliferation" = "#fb9a99", 
                                  "Nucleolin"     = "black", 
                                  "SC35"          = "#1f78b4")) +

    # Define the manual fill palette (matching the lines)
    scale_fill_manual(name = "Marker", 
                      values = c("Proliferation" = "#fb9a99", 
                                 "Nucleolin"     = "black", 
                                 "SC35"          = "#1f78b4")) +

    labs(x = "Pseudotime", y = "Normalized score") +
    monocle3:::monocle_theme_opts() +
    theme(legend.position = "top",             # Legend on top
          legend.title = element_blank(),      # Remove legend title for cleaner look
          axis.text = element_text(color="black"), 
          axis.ticks = element_line(color="black")) +
          
    # Optional: ensure legend keys are legible (solid lines, no grey background)
    guides(color = guide_legend(override.aes = list(alpha = 1, size = 1.5)),
           fill = "none") + 
    theme(aspect.ratio = 1)
```

# Re-calculate pseudotime
## na.omit matters, keep for now

```r
## Don't collapse by embryo
subsample_cds = function(cds) {
    metadata = as.data.frame(colData(cds))
    count_df = metadata %>% count(Experiment, Embryo, Marker, Phenotype, Bin)

    cells_sampled = c()

    for (i in 1:nrow(count_df)) {
        marker = count_df[i,]$Marker
        phenotype = count_df[i,]$Phenotype
        bin = count_df[i,]$Bin
        experiment = count_df[i,]$Experiment
        embryo = count_df[i,]$Embryo

        n = min(count_df[count_df$Marker == marker & count_df$Phenotype == phenotype & 
                         count_df$Experiment == experiment & count_df$Embryo == embryo,]$n)

        cells = metadata %>% filter(Marker == marker & Phenotype == phenotype & Bin == bin & 
                                    Experiment == experiment & Embryo == embryo) %>% pull(Cell)
        cells_sampled = c(cells_sampled, sample(cells, size=n, replace=F))
    }

    cds_subsampled = cds[, cds$Cell %in% cells_sampled]
    
    return(cds_subsampled)
}

compute_bin_score = function(cds, downsample = TRUE) {
    if (downsample) {
        cds_subsampled = subsample_cds(cds)
    } else {
        cds_subsampled = cds
    }
    
    metadata = as.data.frame(colData(cds_subsampled))

    bin_score_list_cluster = list()

    for (marker in unique(metadata$Marker)) {
        bin_score_list_cluster[[marker]] = list()
        for (phenotype in unique(metadata[metadata$Marker == marker,]$Phenotype)) {
            
            median_df = metadata %>%
                filter(Marker == marker & Phenotype == phenotype) %>%
                group_by(Experiment, Embryo, trajectory_label, Cluster_sub_2, Bin) %>%
                summarise(med_median = median(Bin_Median)) %>%
                as.data.frame() %>%
                tidyr::spread(Bin, med_median, fill = 0) %>%
                dplyr::rename(Q1_median = Q1, Q2_median = Q2, Q3_median = Q3, Q4_median = Q4)
            
            bin_scores = metadata %>%
                filter(Marker == marker & Phenotype == phenotype) %>%
                group_by(Experiment, Embryo, Bin, trajectory_label, Cluster_sub_2) %>%
                count() %>%
                as.data.frame() %>%
                tidyr::spread(Bin, n, fill = 0) %>%
                mutate(n = Q1+Q2+Q3+Q4) %>%
                as.data.frame() %>%
                filter(n >= 30) %>%
                left_join(., median_df, by = c("Experiment", "Embryo", "trajectory_label", "Cluster_sub_2")) %>%
                rowwise() %>%
                mutate(score = (Q1_median*Q1 + Q2_median*Q2 + Q3_median*Q3 + Q4_median*Q4) / n) %>%
                # FIXED: Removed the premature group_by and median calculation here!
                as.data.frame()

            bin_scores = bin_scores %>%
                            mutate(Marker = marker,
                                   Phenotype = phenotype)

            bin_score_list_cluster[[marker]][[phenotype]] = bin_scores
        }
    }
    
    bin_score_concat = NULL
    for (marker in names(bin_score_list_cluster)) {
        df = do.call("rbind", bin_score_list_cluster[[marker]])
        if (is.null(bin_score_concat)) {
            bin_score_concat = df
        } else {
            bin_score_concat = rbind(bin_score_concat, df)
        }
    }

    metadata = as.data.frame(colData(cds))
    metadata = left_join(metadata,
          bin_score_concat %>%
              select(-c(Q1, Q2, Q3, Q4, Q1_median, Q2_median, Q3_median, Q4_median, n)), 
          by = c("Experiment", "Embryo", "trajectory_label", "Cluster_sub_2", "Marker", "Phenotype")) %>%
         filter(!is.na(score))
    
    cds = cds[, cds$Cell %in% metadata$Cell]
    cds$Score = metadata[match(cds$Cell, metadata$Cell), ]$score
    cds$Timepoint = as.factor(sapply(cds$Embryo, function(x) strsplit(x, '-')[[1]][1]))
    
    num_cells_expressed_cutoff = round(nrow(as.data.frame(colData(cds)))*0.01)

    cds = detect_genes(cds)
    cds = cds[rowData(cds)$num_cells_expressed > num_cells_expressed_cutoff, ]
    
    return(cds)
}

scale_norm = function(score) {
    (score - min(score))/(max(score)-min(score))
}

get_average_bin_scores = function(cds, downsample = TRUE, n = 30) {
    scores = NULL
    for (i in 1:n) {
        print(i)
        flush.console()
        cds_scores = compute_bin_score(cds, downsample)
        score_df = as.data.frame(colData(cds_scores)) %>%
                        distinct(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2, Score, .keep_all=T) %>%
                        mutate(i = i)
        scores = if (is.null(scores)) {
            scores = score_df
        } else {
            scores = rbind(scores, score_df)
        } 
    }

    scores = scores %>%
        group_by(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2) %>%
        mutate(
            avg_score = mean(Score, na.rm = TRUE), 
            med_score = median(Score, na.rm = TRUE), 
            sd_score  = sd(Score, na.rm = TRUE)
        ) %>%
        as.data.frame() %>%
        # If sd_score is NA (e.g., because there was only 1 valid score to calculate variance), set to 0
        mutate(sd_score = ifelse(is.na(sd_score), 0, sd_score)) %>%
        distinct(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2, .keep_all=T) %>%
        select(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2, avg_score, med_score, sd_score, -i)

    metadata = as.data.frame(colData(cds))
    
    # Kept this safety check so you don't get duplicate columns if you run it twice
    metadata = metadata %>%
        dplyr::select(-dplyr::any_of(c("avg_score", "med_score", "sd_score", "scaled_score", "scaled_score_trajectory", "Score")))
        
    metadata = left_join(metadata, scores, 
          by = c("Experiment", "Embryo", "trajectory_label", "Cluster_sub_2", "Marker", "Phenotype")) 
    
    scaled_scores = metadata %>%
                        distinct(trajectory_label, Cluster_sub_2, Marker, Phenotype, .keep_all=T) %>%
                        group_by(Marker, Phenotype) %>%
                        mutate(scaled_score = scale_norm(avg_score)) %>%
                        select(trajectory_label, Cluster_sub_2, Marker, Phenotype, scaled_score) %>%
                        as.data.frame()

    metadata = metadata %>%
        left_join(., scaled_scores, by = c("trajectory_label", "Cluster_sub_2", "Marker", "Phenotype"))
    
    scaled_scores = metadata %>%
                        distinct(trajectory_label, Cluster_sub_2, Marker, Phenotype, .keep_all=T) %>%
                        group_by(trajectory_label, Marker, Phenotype) %>%
                        mutate(scaled_score_trajectory = scale_norm(avg_score)) %>%
                        select(trajectory_label, Cluster_sub_2, Marker, Phenotype, scaled_score_trajectory) %>%
                        as.data.frame()
    
    metadata = metadata %>%
        left_join(., scaled_scores, by = c("trajectory_label", "Cluster_sub_2", "Marker", "Phenotype")) %>%
        mutate(scaled_score_trajectory = ifelse(is.na(scaled_score_trajectory), 
                                                scaled_score, scaled_score_trajectory))

    
    cds = cds[, cds$Cell %in% metadata$Cell]
    cds$avg_score = metadata[match(cds$Cell, metadata$Cell), ]$avg_score
    cds$med_score = metadata[match(cds$Cell, metadata$Cell), ]$med_score
    cds$sd_score = metadata[match(cds$Cell, metadata$Cell), ]$sd_score
    cds$scaled_score = metadata[match(cds$Cell, metadata$Cell), ]$scaled_score
    cds$scaled_score_trajectory = metadata[match(cds$Cell, metadata$Cell), ]$scaled_score_trajectory
    cds$Timepoint = as.factor(sapply(cds$Embryo, function(x) strsplit(x, '-')[[1]][1]))
    
    num_cells_expressed_cutoff = round(nrow(as.data.frame(colData(cds)))*0.01)

    cds = detect_genes(cds)
    #cds = cds[rowData(cds)$num_cells_expressed > num_cells_expressed_cutoff, ]
                                         
    return(cds)
}
```

```r
## Don't collapse by embryo
subsample_cds = function(cds) {
    metadata = as.data.frame(colData(cds))
    count_df = metadata %>% count(Experiment, Embryo, Marker, Phenotype, Bin)

    cells_sampled = c()

    for (i in 1:nrow(count_df)) {
        marker = count_df[i,]$Marker
        phenotype = count_df[i,]$Phenotype
        bin = count_df[i,]$Bin
        experiment = count_df[i,]$Experiment
        embryo = count_df[i,]$Embryo

        n = min(count_df[count_df$Marker == marker & count_df$Phenotype == phenotype & 
                         count_df$Experiment == experiment & count_df$Embryo == embryo,]$n)

        cells = metadata %>% filter(Marker == marker & Phenotype == phenotype & Bin == bin & 
                                    Experiment == experiment & Embryo == embryo) %>% pull(Cell)
        cells_sampled = c(cells_sampled, sample(cells, size=n, replace=F))
    }

    cds_subsampled = cds[, cds$Cell %in% cells_sampled]
    
    return(cds_subsampled)
}

compute_bin_score = function(cds, downsample = TRUE) {
    if (downsample) {
        cds_subsampled = subsample_cds(cds)
    } else {
        cds_subsampled = cds
    }
    
    metadata = as.data.frame(colData(cds_subsampled))

    bin_score_list_cluster = list()

    for (marker in unique(metadata$Marker)) {
        bin_score_list_cluster[[marker]] = list()
        for (phenotype in unique(metadata[metadata$Marker == marker,]$Phenotype)) {
            
            median_df = metadata %>%
                filter(Marker == marker & Phenotype == phenotype) %>%
                group_by(Experiment, Embryo, trajectory_label, Cluster_sub_2, Bin) %>%
                summarise(med_median = median(Bin_Median)) %>%
                as.data.frame() %>%
                tidyr::spread(Bin, med_median, fill = 0) %>%
                dplyr::rename(Q1_median = Q1, Q2_median = Q2, Q3_median = Q3, Q4_median = Q4)
            
            bin_scores = metadata %>%
                filter(Marker == marker & Phenotype == phenotype) %>%
                group_by(Experiment, Embryo, Bin, trajectory_label, Cluster_sub_2) %>%
                count() %>%
                as.data.frame() %>%
                tidyr::spread(Bin, n, fill = 0) %>%
                mutate(n = Q1+Q2+Q3+Q4) %>%
                as.data.frame() %>%
                filter(n >= 30) %>%
                left_join(., median_df, by = c("Experiment", "Embryo", "trajectory_label", "Cluster_sub_2")) %>%
                rowwise() %>%
                mutate(score = (Q1_median*Q1 + Q2_median*Q2 + Q3_median*Q3 + Q4_median*Q4) / n) %>%
                # FIXED: Removed the premature group_by and median calculation here!
                as.data.frame()

            bin_scores = bin_scores %>%
                            mutate(Marker = marker,
                                   Phenotype = phenotype)

            bin_score_list_cluster[[marker]][[phenotype]] = bin_scores
        }
    }
    
    bin_score_concat = NULL
    for (marker in names(bin_score_list_cluster)) {
        df = do.call("rbind", bin_score_list_cluster[[marker]])
        if (is.null(bin_score_concat)) {
            bin_score_concat = df
        } else {
            bin_score_concat = rbind(bin_score_concat, df)
        }
    }

    metadata = as.data.frame(colData(cds))
    metadata = left_join(metadata,
          bin_score_concat %>%
              select(-c(Q1, Q2, Q3, Q4, Q1_median, Q2_median, Q3_median, Q4_median, n)), 
          by = c("Experiment", "Embryo", "trajectory_label", "Cluster_sub_2", "Marker", "Phenotype")) %>%
    na.omit()
    
    cds = cds[, cds$Cell %in% metadata$Cell]
    cds$Score = metadata[match(cds$Cell, metadata$Cell), ]$score
    cds$Timepoint = as.factor(sapply(cds$Embryo, function(x) strsplit(x, '-')[[1]][1]))
    
    num_cells_expressed_cutoff = round(nrow(as.data.frame(colData(cds)))*0.01)

    cds = detect_genes(cds)
    cds = cds[rowData(cds)$num_cells_expressed > num_cells_expressed_cutoff, ]
    
    return(cds)
}

scale_norm = function(score) {
    (score - min(score))/(max(score)-min(score))
}

get_average_bin_scores = function(cds, downsample = TRUE, n = 30) {
    scores = NULL
    for (i in 1:n) {
        print(i)
        flush.console()
        cds_scores = compute_bin_score(cds, downsample)
        score_df = as.data.frame(colData(cds_scores)) %>%
                        distinct(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2, Score, .keep_all=T) %>%
                        mutate(i = i)
        scores = if (is.null(scores)) {
            scores = score_df
        } else {
            scores = rbind(scores, score_df)
        } 
    }

    # THIS handles your median across iterations perfectly at the Embryo level!
    scores = scores %>%
        group_by(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2) %>%
        mutate(avg_score = mean(Score), med_score = median(Score), sd_score = sd(Score)) %>%
        as.data.frame() %>%
        mutate(sd_score = ifelse(is.na(sd_score), 0, sd_score)) %>%
        distinct(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2, .keep_all=T) %>%
        select(Experiment, Embryo, Marker, Phenotype, trajectory_label, Cluster_sub_2, avg_score, med_score, sd_score, -i)

    metadata = as.data.frame(colData(cds))
    
    # Kept this safety check so you don't get duplicate columns if you run it twice
    metadata = metadata %>%
        dplyr::select(-dplyr::any_of(c("avg_score", "med_score", "sd_score", "scaled_score", "scaled_score_trajectory", "Score")))
        
    metadata = left_join(metadata, scores, 
          by = c("Experiment", "Embryo", "trajectory_label", "Cluster_sub_2", "Marker", "Phenotype")) %>%
        na.omit()
    
    scaled_scores = metadata %>%
                        distinct(trajectory_label, Cluster_sub_2, Marker, Phenotype, .keep_all=T) %>%
                        group_by(Marker, Phenotype) %>%
                        mutate(scaled_score = scale_norm(avg_score)) %>%
                        select(trajectory_label, Cluster_sub_2, Marker, Phenotype, scaled_score) %>%
                        as.data.frame()

    metadata = metadata %>%
        left_join(., scaled_scores, by = c("trajectory_label", "Cluster_sub_2", "Marker", "Phenotype"))
    
    scaled_scores = metadata %>%
                        distinct(trajectory_label, Cluster_sub_2, Marker, Phenotype, .keep_all=T) %>%
                        group_by(trajectory_label, Marker, Phenotype) %>%
                        mutate(scaled_score_trajectory = scale_norm(avg_score)) %>%
                        select(trajectory_label, Cluster_sub_2, Marker, Phenotype, scaled_score_trajectory) %>%
                        as.data.frame()
    
    metadata = metadata %>%
        left_join(., scaled_scores, by = c("trajectory_label", "Cluster_sub_2", "Marker", "Phenotype")) %>%
        mutate(scaled_score_trajectory = ifelse(is.na(scaled_score_trajectory), 
                                                scaled_score, scaled_score_trajectory))

    
    cds = cds[, cds$Cell %in% metadata$Cell]
    cds$avg_score = metadata[match(cds$Cell, metadata$Cell), ]$avg_score
    cds$med_score = metadata[match(cds$Cell, metadata$Cell), ]$med_score
    cds$sd_score = metadata[match(cds$Cell, metadata$Cell), ]$sd_score
    cds$scaled_score = metadata[match(cds$Cell, metadata$Cell), ]$scaled_score
    cds$scaled_score_trajectory = metadata[match(cds$Cell, metadata$Cell), ]$scaled_score_trajectory
    cds$Timepoint = as.factor(sapply(cds$Embryo, function(x) strsplit(x, '-')[[1]][1]))
    
    num_cells_expressed_cutoff = round(nrow(as.data.frame(colData(cds)))*0.01)

    cds = detect_genes(cds)
    #cds = cds[rowData(cds)$num_cells_expressed > num_cells_expressed_cutoff, ]
                                         
    return(cds)
}
```
