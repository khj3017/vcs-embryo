```r
estimate_cell_cycle = function(cds, g1s_markers, g2m_markers){
  cds_g1s = cds[rowData(cds)$gene_short_name %in% g1s_markers,] 
  aggregate_g1s_expression = counts(cds_g1s)
  aggregate_g1s_expression = t(t(aggregate_g1s_expression) / colData(cds_g1s)$Size_Factor)
  aggregate_g1s_expression = Matrix::colSums(aggregate_g1s_expression)
  
  cds_g2m = cds[rowData(cds)$gene_short_name %in% g2m_markers,] 
  aggregate_g2m_expression = counts(cds_g2m)
  aggregate_g2m_expression = t(t(aggregate_g2m_expression) / colData(cds_g2m)$Size_Factor)
  aggregate_g2m_expression = Matrix::colSums(aggregate_g2m_expression)
  #pData(cds)$cell_cycle_score = aggregate_g1s_expression
  cds$g1s_score = log(aggregate_g1s_expression+1)
  cds$g2m_score = log(aggregate_g2m_expression+1)
  cds$proliferation_index = log(aggregate_g1s_expression + aggregate_g2m_expression + 1)
  return(cds)
}

aggregate_gene_score = function(cds, gene_list=c("CCNB1", "CCNB2","CDK1"), name = "gene_list_score"){
  cds_gene_list_norm = cds[rowData(cds)$gene_short_name %in% gene_list,] 
  aggregate_marker_expression = counts(cds_gene_list_norm)
  aggregate_marker_expression = t(t(aggregate_marker_expression) / colData(cds_gene_list_norm)$Size_Factor)
  aggregate_marker_expression = Matrix::colSums(aggregate_marker_expression)
  aggregate_marker_expression = log(aggregate_marker_expression + 1)
  #aggregate_marker_expression = t(scale(t(aggregate_marker_expression)))
  #aggregate_marker_expression = aggregate_marker_expression[rowSums(is.na(aggregate_marker_expression)) == 0,]
  #aggregate_marker_expression = colMeans(aggregate_marker_expression)
  colData(cds)[[name]] = aggregate_marker_expression
  return(cds)
}

normalize <- function(x) {
    return((x- min(x)) /(max(x)-min(x)))
}
```

```r
gene_set_BP = msigdbr::msigdbr(species = "human", subcategory = "GO:BP")
gene_set_hallmarks = msigdbr::msigdbr(species = "human", category = "H")

gene_set_df_BP = gene_set_BP %>% 
                dplyr::distinct(gs_name, gene_short_name=gene_symbol) %>% 
                as.data.frame()

gene_set_hallmarks_df = gene_set_hallmarks %>% 
                dplyr::distinct(gs_name, gene_short_name=gene_symbol) %>% 
                as.data.frame()

granule_genes = gene_set_df_BP %>%
    filter(gs_name == "GOBP_STRESS_GRANULE_ASSEMBLY")
atf6_genes = gene_set_df_BP %>%
    filter(grepl("GOBP_ATF6", gs_name)) %>%
    pull(gene_short_name)
ire1_genes = gene_set_df_BP %>%
    filter(grepl("GOBP_IRE1", gs_name)) %>%
    pull(gene_short_name)
perk_genes = gene_set_df_BP %>%
    filter(grepl("GOBP_PERK", gs_name)) %>%
    pull(gene_short_name)
upr_genes = gene_set_hallmarks_df %>%
    filter(grepl("UNFOLDED", gs_name)) %>%
    pull(gene_short_name)
```

```r
cds_names = list.files(path = '../human-fetal-atlas', pattern = 'cds_*', full.names = T)[-6]

df_all = NULL

for (cds_name in cds_names) {
    
    cds = readRDS(cds_name)
    cds = cluster_cells(cds, resolution = 1e-4)
    cds$Cluster = clusters(cds)
    
    
    cds = estimate_cell_cycle(cds, 
                               g1s_markers = cc.genes$s.genes, 
                               g2m_markers = cc.genes$g2m.genes)
    cds = aggregate_gene_score(cds, granule_genes$gene_short_name, "granule_score")
    cds = aggregate_gene_score(cds, atf6_genes, "atf6_upr_score")
    cds = aggregate_gene_score(cds, ire1_genes, "ire1_upr_score")
    cds = aggregate_gene_score(cds, perk_genes, "perk_upr_score")
    cds = aggregate_gene_score(cds, upr_genes, "upr_score")
    
    cds = aggregate_gene_score(cds, gene_df_0.75_0.3$ortholog_gene, 
                               "nuc_score_0.75_0.3")
    cds = aggregate_gene_score(cds, gene_df_0.7_0.4$ortholog_gene, 
                               "nuc_score_0.7_0.4")

    
    metadata = as.data.frame(colData(cds))
    
    umap_coordinates = reducedDims(cds)[['UMAP']]
    
    metadata = metadata %>%
                 mutate(UMAP1 = umap_coordinates[,1],
                        UMAP2 = umap_coordinates[,2], 
                        UMAP3 = umap_coordinates[,3])
    
    if (is.null(df_all)) {
        df_all = metadata
    } else {
        df_all = rbind(df_all, metadata)
    }

    print(paste0(cds_name, " done"))
    flush.console()
    gc()
    
}
```

```r
set_names <- c("0.75_0.3", "0.7_0.4") 

# 2. Define the normalize function (ensure this is in your environment)
normalize <- function(x) { (x - min(x, na.rm=TRUE)) / (max(x, na.rm=TRUE) - min(x, na.rm=TRUE)) }

for (s_name in set_names) {
    
    model_file <- paste0("nucleolar_model_glmm_Z5_", s_name, "_n_6.RDS")
    raw_score_col  <- paste0("nuc_score_", s_name) 
    norm_score_col <- paste0("normalized_nuc_score_", s_name)
    pred_col       <- paste0("score_pred_glmm_", s_name)
    
    if (file.exists(model_file)) {
        cat("\n[Predicting]", s_name, "...")
        
        # Load the model
        temp_model <- readRDS(model_file)

        df_all[[norm_score_col]] <- normalize(df_all[[raw_score_col]])

        target_var <- all.vars(formula(temp_model))[2] 
        
        pred_df <- data.frame(df_all[[norm_score_col]])
        colnames(pred_df) <- target_var
        
        # C. Generate Predictions
        # re.form = NA means we are predicting the "fixed effects" only (the global trend)
        df_all[[pred_col]] <- as.double(predict(temp_model, 
                                                newdata = pred_df, 
                                                re.form = NA, 
                                                type = "response"))
        
        rm(temp_model)
        cat(" Done.")
    } else {
        warning(paste("Model file not found for:", s_name))
    }
}

gc()
```

```r
margin_spacer <- function(x) {
  # where x is the column in your dataset
  left_length <- nchar(levels(factor(x)))[1]
  if (left_length > 8) {
    return((left_length - 4) * 4)
  }
  else
    return(0)
}
```

```r
library(dplyr)
library(ggplot2)
library(scales)
library(patchwork) # Essential for combining plots with different scales

# 1. Define a function to create a single clean UMAP per organ
create_organ_plot <- function(organ_name, data) {
  
  df_sub <- data %>% filter(Organ == organ_name)
  
  # Calculate local quantiles for this specific organ
  min_val <- quantile(df_sub$score_pred_glmm_0.7_0.4, 0.08, na.rm = TRUE)
  max_val <- quantile(df_sub$score_pred_glmm_0.7_0.4, 0.92, na.rm = TRUE)
  mid_val <- (min_val + max_val) / 2
  
  ggplot(df_sub, aes(x = Main_cluster_umap_1, y = Main_cluster_umap_2)) +
    geom_point(aes(color = score_pred_glmm_0.7_0.4), 
               size = 0.4, 
               stroke = 0.2, 
               alpha = 0.7) +
    
    # Local independent scale
    scale_colour_gradient2(
      low = muted("blue"), 
      mid = "grey90", 
      high = muted("red"), 
      midpoint = mid_val,
      limits = c(min_val, max_val), 
      oob = scales::squish,
      # Format the legend bar to be small
      guide = guide_colorbar(title = "Nucleolar\nfraction", barwidth = 0.5, barheight = 5)
    ) +
    
    labs(title = organ_name) +
    
    #monocle3:::monocle_theme_opts() +
    theme(
        aspect.ratio = 1,
        panel.background = element_rect(fill = "white", color = NA),
        plot.background = element_rect(fill = "white", color = NA),
        axis.title = element_blank(),
        axis.text  = element_blank(),
        axis.ticks = element_blank(),
        axis.line  = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        # Legend styling to keep it from cluttering the plot
        legend.text = element_text(size = 10),
        plot.title = element_text(size = 14, hjust = 0.5)
    )
}

# 2. Generate the list of plots
organs <- unique(df_all$Organ[!is.na(df_all$Organ)])
plot_list <- lapply(organs, create_organ_plot, data = df_all)

# 3. Combine them using patchwork
# 'ncol' sets the grid layout, 'guides = "collect"' would merge legends, 
# so we keep it default to allow individual legends.
options(repr.plot.width = 12, repr.plot.height = 16, repr.plot.res = 300)

wrap_plots(plot_list, ncol = 3)
```

```r
for (ctype in multiple_celltypes) {
    
    df_sub = df_all %>%
        mutate(Organ_cell_lineage = factor(Organ_cell_lineage, 
                            levels = df_ctype_organ$Organ_cell_lineage)) %>%
        filter(!is.na(Organ_cell_lineage)) %>% 
        filter(grepl(ctype, Organ_cell_lineage))
    
    # 1. Order organs by median score
    organ_ordered = df_sub %>%
        group_by(Organ) %>%
        summarise(med_score = median(score_pred_glmm_0.7_0.4, na.rm = TRUE)) %>%
        arrange(med_score) %>%
        pull(Organ)

    # 2. COMPUTE DYNAMIC LIMITS
    # We find the whiskers for every organ and find the global min/max
    whisker_stats <- df_sub %>%
        group_by(Organ) %>%
        summarise(
            low = boxplot.stats(score_pred_glmm_0.7_0.4)$stats[1], # Lower whisker
            high = boxplot.stats(score_pred_glmm_0.7_0.4)$stats[5] # Upper whisker
        )
    
    # Define ylim based on the range of all whiskers with a small (5%) buffer
    # This captures the "body" of every violin and boxplot while hiding outliers
    global_low <- min(whisker_stats$low, na.rm = TRUE)
    global_high <- max(whisker_stats$high, na.rm = TRUE)
    
    # Apply buffer: subtract 5% from min and add 5% to max
    ylim_final <- c(global_low - (0.05 * abs(global_low)), 
                    global_high + (0.05 * abs(global_high)))

    options(repr.plot.width=2.75, repr.plot.height=3.5, repr.plot.res=300)
    g = df_sub %>%
        mutate(Organ = factor(Organ, levels = organ_ordered)) %>%
        ggplot(., aes(x = Organ, y = score_pred_glmm_0.7_0.4)) + 
            geom_violin(aes(fill = Organ), alpha = 0.7, adjust = 2, width = 0.92) + 
            geom_boxplot(width = 0.3, outlier.shape = NA, fatten = 0.75, lwd = 0.75) + 
            labs(x = "", y = "Nucleolar fraction score", title = str_wrap(ctype, width = 20)) +
            monocle3:::monocle_theme_opts() +
            # Use the dynamically calculated global limits
            coord_flip(ylim = ylim_final) + 
            theme(legend.position = "none",
                  axis.title = element_text(color = "black", size = 11),
                  plot.title = element_text(color = "black", size = 11, hjust = 0.5), 
                  axis.text.x = element_text(color = "black", size = 9.5),
                  axis.line = element_line(color = "black"), 
                  axis.text.y = element_text(color = "black", size = 10.5),
                  axis.ticks = element_line(color = "black"),
                  plot.margin = margin(t = 10, r = 10, b = 25, l = 10))
    
    print(g)
    flush.console()
}
```

```r
library(dplyr)

# 1. Summarize UPR Score
df_upr <- df_all %>%
    filter(!is.na(Organ_cell_lineage) & upr_score > 0) %>%
    # First level grouping (Fetus level)
    group_by(Organ, Organ_cell_lineage, Fetus_id, Development_day) %>%
    summarise(med_score = median(score_pred_glmm_0.7_0.4),
              med_target = median(upr_score),
              n = n(), 
              .groups = "drop") %>%
    filter(n >= 50) %>%
    # Second level grouping (Organ/Lineage level)
    group_by(Organ, Organ_cell_lineage) %>%
    summarise(med_score = median(med_score),
              med_target = median(med_target),
              n = n(), 
              .groups = "drop") %>%
    mutate(score_type = "UPR")

# 2. Summarize Granule Score
df_granule <- df_all %>%
    filter(!is.na(Organ_cell_lineage) & granule_score > 0) %>%
    # First level grouping (Fetus level)
    group_by(Organ, Organ_cell_lineage, Fetus_id, Development_day) %>%
    summarise(med_score = median(score_pred_glmm_0.7_0.4),
              med_target = median(granule_score),
              n = n(), 
              .groups = "drop") %>%
    filter(n >= 50) %>%
    # Second level grouping (Organ/Lineage level)
    group_by(Organ, Organ_cell_lineage) %>%
    summarise(med_score = median(med_score),
              med_target = median(med_target),
              n = n(), 
              .groups = "drop") %>%
    mutate(score_type = "Stress granule")

# 3. Combine into one dataframe
df_combined <- bind_rows(df_upr, df_granule)
```

```r
library(dplyr)
library(purrr)


n_filter = 50
safe_cor_test <- possibly(cor.test, otherwise = NULL)

# 2. Create a helper function so we don't have to copy/paste the same code 3 times
calculate_tissue_cor <- function(data, target_col, score_name) {
    # Convert the column name string to a symbol so dplyr can use it
    target_sym <- sym(target_col)
    
    data %>%
        filter(!is.na(Organ_cell_lineage)) %>%
        filter(!!target_sym > 0) %>%
        group_by(Organ, Organ_cell_lineage, Fetus_id) %>%
        summarise(
            med_score = median(score_pred_glmm_0.7_0.4),
            med_target = median(!!target_sym),
            n = n(),
            .groups = "drop"
        ) %>%
        filter(n >= n_filter) %>%
        group_by(Organ, Organ_cell_lineage) %>%
        summarise(
            med_score = median(med_score),
            med_target = median(med_target),
            .groups = "drop"
        ) %>%
        group_by(Organ) %>%
        filter(n() >= 5) %>% # Need at least 3 points per tissue
        tidyr::nest() %>%
        mutate(
            cor_model = map(data, ~ suppressWarnings(
                safe_cor_test(.x$med_score, .x$med_target, method = "spearman")
            ))
        ) %>%
        filter(!map_lgl(cor_model, is.null)) %>%
        mutate(
            score_type = score_name,
            spearman_rho = map_dbl(cor_model, "estimate"),
            p_value = map_dbl(cor_model, "p.value"),
            n_points = map_int(data, nrow)
        ) %>%
        select(score_type, Organ, spearman_rho, p_value, n_points)
}

# 3. Run the function for all three scores
upr_tissue_cor     <- calculate_tissue_cor(df_all, "upr_score", "UPR")
#ups_tissue_cor     <- calculate_tissue_cor(metadata_remapped_filtered, "ups_score", "UPS Score")
granule_tissue_cor <- calculate_tissue_cor(df_all, "granule_score", "Stress granule")

# 4. Combine them all and sort
combined_tissue_cor <- bind_rows(upr_tissue_cor,granule_tissue_cor) %>%
    arrange(score_type, desc(spearman_rho))

# View the final output
print(combined_tissue_cor)
```
