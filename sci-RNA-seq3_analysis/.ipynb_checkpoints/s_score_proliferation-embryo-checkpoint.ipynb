```r
estimate_cell_cycle = function(cds, g1s_markers, g2m_markers){
  cds_g1s = cds[rowData(cds)$gene_short_name %in% g1s_markers,] 
  aggregate_g1s_expression = counts(cds_g1s)
  aggregate_g1s_expression = t(t(aggregate_g1s_expression) / colData(cds_g1s)$Size_Factor)
  aggregate_g1s_expression = Matrix::colSums(aggregate_g1s_expression)
  
  cds_g2m = cds[rowData(cds)$gene_short_name %in% g2m_markers,] 
  aggregate_g2m_expression = counts(cds_g2m)
  aggregate_g2m_expression = t(t(aggregate_g2m_expression) / colData(cds_g2m)$Size_Factor)
  aggregate_g2m_expression = Matrix::colSums(aggregate_g2m_expression)
  #pData(cds)$cell_cycle_score = aggregate_g1s_expression
  cds$g1s_score = log(aggregate_g1s_expression+1)
  cds$g2m_score = log(aggregate_g2m_expression+1)
  cds$proliferation_index = log(aggregate_g1s_expression + aggregate_g2m_expression + 1)
  return(cds)
}

aggregate_gene_score = function(cds, gene_list=c("CCNB1", "CCNB2","CDK1"), name = "gene_list_score"){
  #gene_list = stringr::str_to_title(gene_list)
  cds_gene_list_norm = cds[rowData(cds)$gene_short_name %in% gene_list,] 
  aggregate_marker_expression = counts(cds_gene_list_norm)
  aggregate_marker_expression = t(t(aggregate_marker_expression) / colData(cds_gene_list_norm)$Size_Factor)
  aggregate_marker_expression = Matrix::colSums(aggregate_marker_expression)
  aggregate_marker_expression = log(aggregate_marker_expression + 1)
  #aggregate_marker_expression = t(scale(t(aggregate_marker_expression)))
  #aggregate_marker_expression = aggregate_marker_expression[rowSums(is.na(aggregate_marker_expression)) == 0,]
  #aggregate_marker_expression = colMeans(aggregate_marker_expression)
  colData(cds)[[name]] = aggregate_marker_expression
  return(cds)
}
```

```r
metadata = metadata %>%
    filter((Marker != "SiRDNA" & Phenotype == "Area")) %>%
    group_by(Marker, Phenotype) %>%
    mutate(min_score = quantile(avg_score, 0.025),
           max_score = quantile(avg_score, 0.975)) %>%
    rowwise() %>%
    mutate(norm_score = if (avg_score < min_score) {
        0.025
    } else if (avg_score > max_score) { 
        0.975
    } else {
        (avg_score - min_score) / (max_score - min_score)
    }) %>%
    as.data.frame()
```

```r
corr_pearson_df = as.data.frame(colData(cds_VCS)) %>%
    filter(Phenotype == "Area" & Marker %in% c("Nucleolin", "SC35", "Nuclear")) %>%
    group_by(trajectory_refined, Embryo, Cluster_sub, Marker) %>%
    summarise(med_score = median(avg_score), med_prolif = median(proliferation_index), .groups = "drop") %>%
    mutate(Embryo = as.factor(Embryo)) %>%
    group_by(trajectory_refined, Marker) %>% 
    add_tally() %>% 
    filter(n >= 10) %>%
    
    group_by(trajectory_refined) %>%
    mutate(prolif_90 = quantile(med_prolif, 0.9)) %>% 
    filter(prolif_90 > 1.2) %>%
    
    group_by(trajectory_refined, Marker) %>%
    
    filter(n_distinct(Embryo) > 1) %>%
    
    summarise(
        # pick(everything()) passes the current group's data into the rmcorr model
        rmc_r = rmcorr(participant = Embryo, measure1 = med_score, measure2 = med_prolif, dataset = pick(everything()))$r,
        rmc_p = rmcorr(participant = Embryo, measure1 = med_score, measure2 = med_prolif, dataset = pick(everything()))$p,
        .groups = "drop"
    )

corr_pearson_df
```

```r
corr_spearman_df = as.data.frame(colData(cds_VCS)) %>%
    filter(Phenotype == "Area" & Marker %in% c("Nucleolin", "SC35", "Nuclear")) %>%
    group_by(trajectory_refined, Embryo, Cluster_sub, Marker) %>%
    summarise(med_score = median(avg_score), med_prolif = median(proliferation_index), .groups = "drop") %>%
    
    mutate(Embryo = as.factor(Embryo)) %>%
    
    group_by(trajectory_refined, Marker) %>% 
    add_tally() %>% 
    filter(n>=10) %>%
    
    group_by(trajectory_refined) %>%
    mutate(prolif_90 = quantile(med_prolif, 0.9)) %>% 
    filter(prolif_90 > 1.2) %>%
    
    group_by(trajectory_refined, Marker) %>%
    filter(n_distinct(Embryo) > 1) %>%
    
    mutate(
        ranked_score = rank(med_score, ties.method = "average"),
        ranked_prolif = rank(med_prolif, ties.method = "average")
    ) %>%

    summarise(
        rmc_r = rmcorr(participant = Embryo, measure1 = ranked_score, measure2 = ranked_prolif, dataset = pick(everything()))$r,
        rmc_p = rmcorr(participant = Embryo, measure1 = ranked_score, measure2 = ranked_prolif, dataset = pick(everything()))$p,
        .groups = "drop"
    )

corr_spearman_df

```

```r
options(repr.plot.width=2.7, repr.plot.height=3.5, repr.plot.res=300)
corr_spearman_df %>% 
    ggplot(., aes(x = Marker, y = rmc_r)) + 
        geom_boxplot(aes(fill = Marker), lwd=0.8, fatten=0.8, outlier.shape = NA) + 
        geom_jitter(size = 1.2) + 
        scale_fill_manual(values = c("#B668A1", "#DC9200", "#6baed6")) + 
        # PLOTTING: Updated label to indicate this is a Ranked (Spearman-equivalent) correlation
        labs(x = "", y = "Spearman correlation\n(Size score vs proliferation index)", fill = "Marker") +  
        monocle3:::monocle_theme_opts() + 
        theme(axis.text.x = element_text(color="black", size = 12), 
              axis.text.y = element_text(color = "black", size = 12),
              axis.ticks = element_line(color = "black"),
              legend.position = "none")
```
