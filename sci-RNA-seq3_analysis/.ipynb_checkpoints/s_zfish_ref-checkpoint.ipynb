```r
aggregate_gene_score = function(cds, gene_list=c("CCNB1", "CCNB2","CDK1"), name = "gene_list_score"){
  cds_gene_list_norm = cds[rowData(cds)$gene_short_name %in% gene_list,] 
  aggregate_marker_expression = counts(cds_gene_list_norm)
  aggregate_marker_expression = t(t(aggregate_marker_expression) / colData(cds_gene_list_norm)$Size_Factor)
  aggregate_marker_expression = Matrix::colSums(aggregate_marker_expression)
  aggregate_marker_expression = log(aggregate_marker_expression + 1)
  #aggregate_marker_expression = t(scale(t(aggregate_marker_expression)))
  #aggregate_marker_expression = aggregate_marker_expression[rowSums(is.na(aggregate_marker_expression)) == 0,]
  #aggregate_marker_expression = colMeans(aggregate_marker_expression)
  colData(cds)[[name]] = aggregate_marker_expression
  return(cds)
}

estimate_cell_cycle = function(cds, g1s_markers, g2m_markers){
  cds_g1s = cds[rowData(cds)$gene_short_name %in% g1s_markers,] 
  aggregate_g1s_expression = counts(cds_g1s)
  aggregate_g1s_expression = t(t(aggregate_g1s_expression) / colData(cds_g1s)$Size_Factor)
  aggregate_g1s_expression = Matrix::colSums(aggregate_g1s_expression)
  
  cds_g2m = cds[rowData(cds)$gene_short_name %in% g2m_markers,] 
  aggregate_g2m_expression = counts(cds_g2m)
  aggregate_g2m_expression = t(t(aggregate_g2m_expression) / colData(cds_g2m)$Size_Factor)
  aggregate_g2m_expression = Matrix::colSums(aggregate_g2m_expression)
  #pData(cds)$cell_cycle_score = aggregate_g1s_expression
  cds$g1s_score = log(aggregate_g1s_expression+1)
  cds$g2m_score = log(aggregate_g2m_expression+1)
  cds$proliferation_index = log(aggregate_g1s_expression + aggregate_g2m_expression + 1)
  return(cds)
}

pseudobulk_cds = function(cds_sub, sample_group = "Cluster_group", cell_group = "Grouping") {
    cf_ccs = new_cell_count_set(cds_sub,
                            sample_group = sample_group,
                            cell_group = cell_group)
    
    pb_cds = hooke:::pseudobulk_ccs_for_states(cf_ccs)
    
    return(pb_cds)
    
}

normalize <- function(x) {
    return((x- min(x)) /(max(x)-min(x)))
}
```

```r
library(dplyr)
library(purrr)

# 1. Create the safe version of cor.test
safe_cor_test <- possibly(cor.test, otherwise = NULL)

# 2. Create a helper function so we don't have to copy/paste the same code 3 times
calculate_tissue_cor <- function(data, target_col, score_name) {
    # Convert the column name string to a symbol so dplyr can use it
    target_sym <- sym(target_col)
    
    data %>%
        # Filter for values > 0 for the specific score
        filter(!!target_sym > 0) %>%
        group_by(tissue, cell_type, embryo_ID) %>%
        summarise(
            med_score = median(score_pred_glmm_0.7_0.4_remapped),
            med_target = median(!!target_sym),
            n = n(),
            .groups = "drop"
        ) %>%
        filter(n >= n_filter) %>%
        group_by(tissue) %>%
        filter(n() >= 5) %>% # Need at least 3 points per tissue
        nest() %>%
        mutate(
            cor_model = map(data, ~ suppressWarnings(
                safe_cor_test(.x$med_score, .x$med_target, method = "spearman")
            ))
        ) %>%
        filter(!map_lgl(cor_model, is.null)) %>%
        mutate(
            score_type = score_name,
            spearman_rho = map_dbl(cor_model, "estimate"),
            p_value = map_dbl(cor_model, "p.value"),
            n_points = map_int(data, nrow)
        ) %>%
        select(score_type, tissue, spearman_rho, p_value, n_points)
}

# 3. Run the function for all three scores
upr_tissue_cor     <- calculate_tissue_cor(metadata_remapped_filtered, "upr_score", "UPR Score")
ups_tissue_cor     <- calculate_tissue_cor(metadata_remapped_filtered, "ups_score", "UPS Score")
granule_tissue_cor <- calculate_tissue_cor(metadata_remapped_filtered, "granule_score", "Granule Score")

# 4. Combine them all and sort
combined_tissue_cor <- bind_rows(upr_tissue_cor, ups_tissue_cor, granule_tissue_cor) %>%
    arrange(score_type, desc(spearman_rho))

# View the final output
print(combined_tissue_cor)
```

```r
library(glmmTMB)
library(dplyr)

# 1. Fit the GLMM
# We use a standard Gaussian family since we are comparing medians.
# (1 | embryo_ID) tells the model to account for baseline differences between embryos.
glmm_model <- glmmTMB(med_granule ~ med_score + (1 | embryo_ID), 
                      data = df_summarised, 
                      family = gaussian())

# 1. Create the dummy dataframe for the x-axis range
curve_data <- data.frame(
    med_score = seq(min(df_summarised$med_score, na.rm = TRUE), 
                    max(df_summarised$med_score, na.rm = TRUE), 
                    length.out = 100)
)

# 2. Predict the fixed effect AND get the standard errors
# se.fit = TRUE generates a list with both the fit and the standard error
preds <- predict(glmm_model, newdata = curve_data, re.form = NA, se.fit = TRUE)

# 3. Extract the fit and calculate the 95% Confidence Interval
curve_data$med_granule <- preds$fit
curve_data$se <- preds$se.fit

# 95% CI is roughly the estimate +/- 1.96 * standard error
curve_data$ci_lower <- curve_data$med_granule - (1.96 * curve_data$se)
curve_data$ci_upper <- curve_data$med_granule + (1.96 * curve_data$se)
```

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res=300)

ggplot(df_summarised, aes(x = med_score, y = med_granule)) + 
    # 1. The data points
    geom_jitter(size = 0.6, width = 0.002, height = 0.005, alpha = 0.7, color = "black") + 
    
    # 2. The shaded variance/confidence interval
    # Note: We specify ymin and ymax here for the ribbon bounds
    geom_ribbon(data = curve_data, 
                aes(x = med_score, y = med_granule, ymin = ci_lower, ymax = ci_upper), 
                alpha = 0.2, fill = "#2166ac") +
    
    # 3. The solid fixed-effect fit line
    geom_line(data = curve_data, aes(x = med_score, y = med_granule), 
              color = "#2166ac", linewidth = 1.2) +
    
    # 4. Formatting and themes
    #xlim(c(0.1, 0.3)) + ylim(c(0.5, 1.22)) + 
    labs(x = "Nucleolar fraction score", y = "Stress granule score") + 
    monocle3:::monocle_theme_opts() + theme(legend.position = "none") +
    theme(axis.text.x = element_text(color = "black", size = 12), 
          axis.text.y = element_text(color = "black", size = 12), 
          strip.text = element_text(color = "black", size = 12),
          axis.ticks = element_line(color = "black"))
```

```r
library(dplyr)
library(purrr)

n_filter = 20
safe_cor_test <- possibly(cor.test, otherwise = NULL)

# 2. Create a helper function so we don't have to copy/paste the same code 3 times
calculate_tissue_cor <- function(data, target_col, score_name) {
    # Convert the column name string to a symbol so dplyr can use it
    target_sym <- sym(target_col)
    
    data %>%
        # Filter for values > 0 for the specific score
        filter(!!target_sym > 0) %>%
        group_by(tissue, cell_type, embryo_ID) %>%
        summarise(
            med_score = median(score_pred_glmm_0.7_0.4_remapped),
            med_target = median(!!target_sym),
            n = n(),
            .groups = "drop"
        ) %>%
        filter(n >= n_filter) %>%
        group_by(tissue, cell_type) %>%
        summarise(
            med_score = median(med_score),
            med_target = median(med_target),
            .groups = "drop"
        ) %>%
        group_by(tissue) %>%
        filter(n() >= 5) %>% # Need at least 3 points per tissue
        nest() %>%
        mutate(
            cor_model = map(data, ~ suppressWarnings(
                safe_cor_test(.x$med_score, .x$med_target, method = "spearman")
            ))
        ) %>%
        filter(!map_lgl(cor_model, is.null)) %>%
        mutate(
            score_type = score_name,
            spearman_rho = map_dbl(cor_model, "estimate"),
            p_value = map_dbl(cor_model, "p.value"),
            n_points = map_int(data, nrow)
        ) %>%
        select(score_type, tissue, spearman_rho, p_value, n_points)
}

# 3. Run the function for all three scores
upr_tissue_cor     <- calculate_tissue_cor(metadata_remapped_filtered, "upr_score", "UPR")
#ups_tissue_cor     <- calculate_tissue_cor(metadata_remapped_filtered, "ups_score", "UPS Score")
granule_tissue_cor <- calculate_tissue_cor(metadata_remapped_filtered, "granule_score", "Stress granule")

# 4. Combine them all and sort
combined_tissue_cor <- bind_rows(upr_tissue_cor,granule_tissue_cor) %>%
    arrange(score_type, desc(spearman_rho))

# View the final output
print(combined_tissue_cor)
```
