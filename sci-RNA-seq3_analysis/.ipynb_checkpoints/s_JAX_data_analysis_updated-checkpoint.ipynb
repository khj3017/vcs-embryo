```r
aggregate_gene_score_mm = function(mm, sf_df){
  aggregate_marker_expression = t(t(mm) / sf_df$Size_Factor)
  aggregate_marker_expression = Matrix::colSums(aggregate_marker_expression)
  aggregate_marker_expression = log10(aggregate_marker_expression + 1)
  sf_df$gene_list_score = aggregate_marker_expression
  return(sf_df)
}

aggregate_gene_score = function(cds, gene_list=c("CCNB1", "CCNB2","CDK1"), name = "gene_list_score"){
  cds_gene_list_norm = cds[rowData(cds)$gene_short_name %in% gene_list,] 
  aggregate_marker_expression = counts(cds_gene_list_norm)
  aggregate_marker_expression = t(t(aggregate_marker_expression) / colData(cds_gene_list_norm)$Size_Factor)
  aggregate_marker_expression = Matrix::colSums(aggregate_marker_expression)
  aggregate_marker_expression = log(aggregate_marker_expression + 1)
  #aggregate_marker_expression = t(scale(t(aggregate_marker_expression)))
  #aggregate_marker_expression = aggregate_marker_expression[rowSums(is.na(aggregate_marker_expression)) == 0,]
  #aggregate_marker_expression = colMeans(aggregate_marker_expression)
  colData(cds)[[name]] = aggregate_marker_expression
  return(cds)
}

estimate_cell_cycle = function(cds, g1s_markers, g2m_markers){
  cds_g1s = cds[rowData(cds)$gene_short_name %in% g1s_markers,] 
  aggregate_g1s_expression = counts(cds_g1s)
  aggregate_g1s_expression = t(t(aggregate_g1s_expression) / colData(cds_g1s)$Size_Factor)
  aggregate_g1s_expression = Matrix::colSums(aggregate_g1s_expression)
  
  cds_g2m = cds[rowData(cds)$gene_short_name %in% g2m_markers,] 
  aggregate_g2m_expression = counts(cds_g2m)
  aggregate_g2m_expression = t(t(aggregate_g2m_expression) / colData(cds_g2m)$Size_Factor)
  aggregate_g2m_expression = Matrix::colSums(aggregate_g2m_expression)
  #pData(cds)$cell_cycle_score = aggregate_g1s_expression
  cds$g1s_score = log(aggregate_g1s_expression+1)
  cds$g2m_score = log(aggregate_g2m_expression+1)
  cds$proliferation_index = log(aggregate_g1s_expression + aggregate_g2m_expression + 1)
  return(cds)
}

normalize <- function(x) {
    return((x- min(x)) /(max(x)-min(x)))
}
```

```r
gene_set_BP = msigdbr::msigdbr(species = "mouse", subcategory = "GO:BP")
gene_set_CC = msigdbr::msigdbr(species = "mouse", subcategory = "GO:CC")
gene_set_hallmarks = msigdbr::msigdbr(species = "mouse", category = "H")
gene_set_reactome = msigdbr::msigdbr(species = "mouse", subcategory = "CP:REACTOME")
gene_set_cgp = msigdbr::msigdbr(species = "mouse", subcategory = "CGP")

gene_set_BP_df = gene_set_BP %>% 
                dplyr::distinct(gs_name, gene_short_name=gene_symbol) %>% 
                as.data.frame()

gene_set_hallmarks_df = gene_set_hallmarks %>% 
                dplyr::distinct(gs_name, gene_short_name=gene_symbol) %>% 
                as.data.frame()

gene_set_reactome_df = gene_set_reactome %>% 
                dplyr::distinct(gs_name, gene_short_name=gene_symbol) %>% 
                as.data.frame()

gene_set_CC_df = gene_set_CC %>% 
                dplyr::distinct(gs_name, gene_short_name=gene_symbol) %>% 
                as.data.frame()
```

```r
granule_genes = gene_set_BP_df %>%
    filter(grepl("GOBP_STRESS_GRANULE_ASSEMBLY", gs_name)) %>%
    pull(gene_short_name)
atf6_genes = gene_set_BP_df %>%
    filter(grepl("GOBP_ATF6", gs_name)) %>%
    pull(gene_short_name)
ire1_genes = gene_set_BP_df %>%
    filter(grepl("GOBP_IRE1", gs_name)) %>%
    pull(gene_short_name)
perk_genes = gene_set_BP_df %>%
    filter(grepl("GOBP_PERK", gs_name)) %>%
    pull(gene_short_name)
upr_genes = gene_set_hallmarks_df %>%
    filter(grepl("UNFOLDED", gs_name)) %>%
    pull(gene_short_name)
dystroglycan_genes = gene_set_CC_df %>%
    filter(gs_name == "GOCC_DYSTROGLYCAN_COMPLEX") %>%
    pull(gene_short_name)
catenin_genes = gene_set_CC_df %>%
    filter(gs_name == "GOCC_CATENIN_COMPLEX") %>%
    pull(gene_short_name)
adhesion_complex_genes = gene_set_CC_df %>%
    filter(gs_name == "GOCC_PROTEIN_COMPLEX_INVOLVED_IN_CELL_CELL_ADHESION") %>%
    pull(gene_short_name)
cpeb_genes = gene_set_CC_df %>%
    filter(gs_name == "GOCC_MESSENGER_RIBONUCLEOPROTEIN_COMPLEX") %>%
    filter(grepl("Cpeb", gene_short_name)) %>%
    pull(gene_short_name)
```

## Estimate nucleolar score

```r
df_cell = readRDS("../E15.25-nucleolin-SiRDNA-SC35-combine-all-cds/df_cell.rds")
df_birth_series = readRDS("../JAX_data/df_cell.birth_series.rds")
df_early = readRDS("../JAX_data/df_cell.somitogenesis_validation.rds")

df_cell = bind_rows(df_cell, df_birth_series, df_early) %>%
             mutate(Size_Factor = UMI_count / exp(mean(log(UMI_count))))

dim(df_cell)

df_gene = read.csv("../JAX_data/GSE228590_gene_annotation.csv")
head(df_gene)
```

```r
gene_files <- list.files(pattern = "nucleolar_genes_Z.*\\.RDS$")

# Initialize the list
nuc_sets_to_score <- list()

# Load them and clean up the names
for (f in gene_files) {
    # Extract the threshold name (everything between 'genes_' and '.RDS')
    set_name <- gsub("nucleolar_genes_|.RDS", "", f)
    nuc_sets_to_score[[set_name]] <- readRDS(f)
    message("Loaded gene set: ", set_name)
}
```

```r
set_names <- names(nuc_sets_to_score) 
nuc_models <- list()

for (s_name in set_names) {
    
    # Define filenames and column names
    model_file <- paste0("nucleolar_model_glmm_", s_name, ".RDS")
    raw_score_col <- paste0("nuc_score_", s_name) # Ensure this matches your loop naming
    norm_score_col <- paste0("normalized_nuc_score_", s_name)
    pred_col <- paste0("score_pred_glmm_", s_name)
    
    if (file.exists(model_file)) {
        cat("\n[Predicting]", s_name, "...")
        
        # A. Load the model
        nuc_models[[s_name]] <- readRDS(model_file)
        message("Loaded model: ", set_name)
    }
}

# Clear garbage
gc()
```

```r
cds_list_names = c("JAX_E8-8.5.RDS", "JAX_E8-9.25.RDS", "JAX_E8.75-9.5.RDS", "JAX_E9.75-10-1.RDS", 
                   "JAX_E9.75-10-2.RDS", "JAX_E10.25-10.75.RDS", "JAX_E11-11.75.RDS", "JAX_E12-12.75.RDS", 
                   "JAX_E13-13.75.RDS", "JAX_E14-14.33.RDS", "JAX_E15-15.75.RDS", "JAX_E16-16.75.RDS", 
                   "JAX_E16-16.75-2.RDS", "JAX_E17-17.75.RDS", "JAX_E18-18.75.RDS", "JAX_E18-18.75-2.RDS", 
                   "JAX_P0-1.RDS", "JAX_P0-2.RDS", "JAX_birth-series.RDS")

df_all = NULL

for (cds_name in cds_list_names) {
    cat("\n[STARTING]", cds_name, "...")
    
    # 1. Load CDS
    cds_path = paste0("../JAX_data/cds_", cds_name)
    if(!file.exists(cds_path)) {
        cat(" File not found. Skipping.\n")
        next
    }
    cds_sub = readRDS(cds_path)
    
    df_sub_meta = df_cell[df_cell$cell_id %in% colData(cds_sub)$cell_id, ]
    colData(cds_sub)$Size_Factor = df_sub_meta$Size_Factor
    monocle3:::size_factors(cds_sub) = colData(cds_sub)$Size_Factor
    
    cds_sub = estimate_cell_cycle(cds_sub, 
                                  g1s_markers = cc.genes$s.genes, 
                                  g2m_markers = cc.genes$g2m.genes)
    
    cds_sub = aggregate_gene_score(cds_sub, granule_genes, "granule_score")
    cds_sub = aggregate_gene_score(cds_sub, atf6_genes, "atf6_upr_score")
    cds_sub = aggregate_gene_score(cds_sub, ire1_genes, "ire1_upr_score")
    cds_sub = aggregate_gene_score(cds_sub, perk_genes, "perk_upr_score")
    cds_sub = aggregate_gene_score(cds_sub, upr_genes, "upr_score")
    cds_sub = aggregate_gene_score(cds_sub, c("Malat1"), "Malat1")
    

    for (set_label in names(nuc_sets_to_score)) {
        genes_to_use <- nuc_sets_to_score[[set_label]]$gene_short_name
        score_name <- paste0("nuc_score_", set_label)
        
        cds_sub = aggregate_gene_score(cds_sub, genes_to_use, score_name)
    }

    df_step = as.data.frame(colData(cds_sub))
    
    if (is.null(df_all)) {
        df_all = df_step
    } else {
        df_all = bind_rows(df_all, df_step)
    }
    
    # 7. Memory Cleanup
    cat(" Complete. Total rows in df_all:", nrow(df_all), "\n")
    rm(cds_sub, df_step)
    flush.console()
    gc() # Garbage collection
}

head(df_all)
```

```r
set_names <- names(nuc_sets_to_score) 

for (s_name in set_names) {
    
    # Define filenames and column names
    model_file <- paste0("nucleolar_model_glmm_", s_name, ".RDS")
    raw_score_col <- paste0("nuc_score_", s_name) # Ensure this matches your loop naming
    norm_score_col <- paste0("normalized_nuc_score_", s_name)
    pred_col <- paste0("score_pred_glmm_", s_name)
    
    if (file.exists(model_file)) {
        cat("\n[Predicting]", s_name, "...")
        
        # A. Load the model
        temp_model <- readRDS(model_file)

        df_all[[norm_score_col]] <- normalize(df_all[[raw_score_col]])
        target_var <- all.vars(formula(temp_model))[2] 
        
        pred_df <- data.frame(df_all[[norm_score_col]])
        colnames(pred_df) <- target_var
        
        df_all[[pred_col]] <- as.double(predict(temp_model, 
                                                newdata = pred_df, 
                                                re.form = NA, 
                                                type = "response"))
        
        # Cleanup memory for the next iteration
        rm(temp_model)
        cat(" Done.")
    } else {
        warning(paste("Model file not found for:", s_name))
    }
}

# Clear garbage
gc()
```

```r
library(dplyr)
library(ggplot2)
library(scales)
library(patchwork) # Essential for combining plots with different scales

# 1. Define a function to create a single clean UMAP per organ
create_trajectory_plot <- function(trajectory, data) {
  
  df_sub <- data %>%
                filter(major_trajectory == trajectory)
  
  # Calculate local quantiles for this specific organ
  min_val <- quantile(df_sub$score_pred_glmm_Z5_0.7_0.4_n_6, 0.08, na.rm = TRUE)
  max_val <- quantile(df_sub$score_pred_glmm_Z5_0.7_0.4_n_6, 0.92, na.rm = TRUE)
  mid_val <- (min_val + max_val) / 2
  
  ggplot(df_sub, aes(x = UMAP_2d_1, y = UMAP_2d_2)) +
    geom_point(aes(color = score_pred_glmm_Z5_0.7_0.4_n_6), 
               size = 0.4, 
               stroke = 0.2, 
               alpha = 0.7) +
    
    # Local independent scale
    scale_colour_gradient2(
      low = muted("blue"), 
      mid = "grey90", 
      high = muted("red"), 
      midpoint = mid_val,
      limits = c(min_val, max_val), 
      oob = scales::squish,
      # Format the legend bar to be small
      guide = guide_colorbar(title = "Nucleolar\nfraction", barwidth = 0.5, barheight = 5)
    ) +
    
    labs(title = str_wrap(str_replace_all(trajectory, "_", " "), width = 15)) + 
    
    #monocle3:::monocle_theme_opts() +
    theme(
        aspect.ratio = 1,
        panel.background = element_rect(fill = "white", color = NA),
        plot.background = element_rect(fill = "white", color = NA),
        axis.title = element_blank(),
        axis.text  = element_blank(),
        axis.ticks = element_blank(),
        axis.line  = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        # Legend styling to keep it from cluttering the plot
        legend.text = element_text(size = 10),
        plot.title = element_text(size = 14, hjust = 0.5)
    )
}

# 2. Generate the list of plots
major_trajectories <- unique(df_embryo$major_trajectory[!is.na(df_embryo$major_trajectory)])
plot_list <- lapply(major_trajectories, create_trajectory_plot, data = df_embryo)

# 3. Combine them using patchwork
# 'ncol' sets the grid layout, 'guides = "collect"' would merge legends, 
# so we keep it default to allow individual legends.
options(repr.plot.width = 12, repr.plot.height = 16, repr.plot.res = 300)

wrap_plots(plot_list, ncol = 4)
```

```r
evaluate_inverse_trend <- function(df) {
  if(nrow(df) < 15) {
    return(tibble(spearman_rho = NA, min_fit = NA, max_fit = NA, 
                  fit_amplitude = NA, percent_change = NA, is_inverse = FALSE))
  }
  
  # 1. Global Trend: Spearman correlation
  # Order doesn't change the rho value, but logically it is now X, Y
  rho <- cor(df$med_prolif, df$med_score_pred_glmm_Z5_0.7_0.4_n_6, 
             method = "spearman", use = "complete.obs")
  
  # 2. Fit the Spline: Score (Y) ~ ns(Proliferation (X))
  fit <- lm(med_score_pred_glmm_Z5_0.7_0.4_n_6 ~ ns(med_prolif, df = 3), data = df)
  
  # Predict the Score (Y) across a dense grid of 100 Proliferation (X) points
  x_range <- range(df$med_prolif, na.rm = TRUE)
  x_grid <- seq(x_range[1], x_range[2], length.out = 100)
  y_pred_grid <- predict(fit, newdata = data.frame(med_prolif = x_grid))
  
  # 3. Compute Min, Max, and Change of the predicted Score
  max_fit <- max(y_pred_grid, na.rm = TRUE)
  min_fit <- min(y_pred_grid, na.rm = TRUE)
  
  # Absolute amplitude between the highest and lowest points of the Score curve
  fit_amplitude <- (max_fit - min_fit) / max_fit 
  
  # Directional logic: Does the Score end lower than it started as Proliferation increases?
  net_direction <- y_pred_grid[100] - y_pred_grid[1] 
  
  # 4. The Filter Logic
  tibble(
    spearman_rho = rho,
    min_fit = min_fit,
    max_fit = max_fit,
    fit_amplitude = fit_amplitude,
    
    # Calculate the drop as a percentage of the peak Score
    percent_change = (min_fit - max_fit) / max_fit * 100, 
    
    # Keep it IF the overall correlation is negative AND the Score drops directionally
    is_inverse = (rho < -0.05) & (net_direction < 0) 
  )
}

# Apply to your data
all_trends_df <- celltype_day_raw %>%
  filter(n >= 300) %>%
  group_by(celltype_update) %>%
  nest() %>%
  mutate(trend_stats = map(data, evaluate_inverse_trend)) %>%
  unnest(trend_stats) %>%
  filter(!is.na(is_inverse))

# View the results sorted by the largest amplitude (most dramatic curve swings)
print(
  all_trends_df %>% 
    arrange(desc(fit_amplitude)) %>% 
    select(celltype_update, spearman_rho, min_fit, max_fit, fit_amplitude, percent_change)
)
```

```r
library(dplyr)
library(tidyr)
library(purrr)
library(splines)

evaluate_inverse_trend <- function(df) {
  if(nrow(df) < 10) {
    return(tibble(percent_drop = NA, prop_down = NA, plunge_rho = NA, is_inverse = FALSE))
  }
  
  # 1. Fit & Predict over a grid
  fit <- lm(med_score_pred_glmm_Z5_0.7_0.4_n_6 ~ ns(med_prolif, df = 3), data = df)
  x_range <- range(df$med_prolif, na.rm = TRUE)
  x_grid <- seq(x_range[1], x_range[2], length.out = 100)
  y_pred <- predict(fit, newdata = data.frame(med_prolif = x_grid))
  
  # 2. Cummax Plunge (Magnitude)
  plunges <- cummax(y_pred) - y_pred
  percent_drop <- max(plunges, na.rm = TRUE) / cummax(y_pred)[which.max(plunges)]
  
  # 3. Proportion Downward
  slopes <- diff(y_pred)
  prop_down <- sum(slopes < 0) / length(slopes)
  
  # 4. NEW: SEGMENTED CORRELATION ("Plunge Rho")
  # Find the exact Proliferation value (X) where the Score hits its minimum (the trough)
  x_trough <- x_grid[which.min(y_pred)]
  
  # Isolate the data points that happen BEFORE the trough
  df_plunge <- df %>% filter(med_prolif <= x_trough)
  
  # Calculate the Spearman correlation for just this downward phase
  # (Requires at least 5 points in the segment to be mathematically stable)
  if(nrow(df_plunge) >= 5) {
    plunge_rho <- round(cor(df_plunge$med_prolif, df_plunge$med_score_pred_glmm_Z5_0.7_0.4_n_6, 
                      method = "spearman", use = "complete.obs"), 3)
  } else {
    plunge_rho <- NA
  }
  
  # 5. THE REFINED FILTER LOGIC
  # - Must drop by at least 2%
  # - Must occupy at least 25% of the curve
  # - The targeted downward segment must have a genuinely convincing negative correlation (< -0.3)
  is_inverse <- (percent_drop >= 0.05) & (prop_down >= 0.5) & (!is.na(plunge_rho)) & (plunge_rho <= -0.3)

  tibble(
    min_fit = min(y_pred, na.rm = TRUE),
    max_fit = max(y_pred, na.rm = TRUE),
    percent_drop = percent_drop,
    prop_down = prop_down,
    plunge_rho = plunge_rho,
    is_inverse = is_inverse
  )
}

# Apply to your data
all_trends_df <- celltype_day_raw %>%
  filter(n >= 100) %>%
  group_by(celltype_update) %>%
  nest() %>%
  mutate(trend_stats = map(data, evaluate_inverse_trend)) %>%
  unnest(trend_stats) %>%
  filter(celltype_update %in% celltype_filter$celltype_update) %>%
  filter(!is.na(is_inverse))

# View the convincing metrics
print(all_trends_df %>% 
        filter(is_inverse == TRUE) %>% 
        select(celltype_update, percent_drop, prop_down, plunge_rho))
```

```r
library(dplyr)
library(tidyr)
library(purrr)
library(splines)

evaluate_inverse_trend <- function(df) {
  if(nrow(df) < 15) {
    return(tibble(percent_drop = NA, prop_down = NA, plunge_rho = NA, is_inverse = FALSE))
  }
  
  # 1. Fit & Predict over a grid
  fit <- lm(med_score_pred_glmm_Z5_0.7_0.4_n_6 ~ ns(med_prolif, df = 3), data = df)
  x_range <- range(df$med_prolif, na.rm = TRUE)
  x_grid <- seq(x_range[1], x_range[2], length.out = 100)
  y_pred <- predict(fit, newdata = data.frame(med_prolif = x_grid))
  
  # 2. Identify the exact boundaries of the biggest Plunge
  plunges <- cummax(y_pred) - y_pred
  idx_trough <- which.max(plunges) # Index of the bottom of the drop
  idx_peak <- which.max(y_pred[1:idx_trough]) # Index of the peak right before the drop
  
  percent_drop <- plunges[idx_trough] / y_pred[idx_peak]
  
  # 3. Proportion Downward
  slopes <- diff(y_pred)
  prop_down <- sum(slopes < 0) / length(slopes)
  
  # 4. NEW: SEGMENTED CORRELATION ("Plunge Rho")
  x_peak <- x_grid[idx_peak]
  x_trough <- x_grid[idx_trough]
  
  # Isolate the data points that fall strictly within the dropping segment
  df_plunge <- df %>% filter(med_prolif >= x_peak & med_prolif <= x_trough)
  
  # Calculate the Spearman correlation for just this downward phase
  if(nrow(df_plunge) >= 5) {
    plunge_rho <- round(cor(df_plunge$med_prolif, df_plunge$med_score_pred_glmm_Z5_0.7_0.4_n_6, 
                            method = "spearman", use = "complete.obs"), 3)
  } else {
    plunge_rho <- NA
  }
  
  # 5. THE REFINED FILTER LOGIC
  # Note: I lowered prop_down to 0.3. In your plot, the upward curve takes up roughly
  # 60% of the X-axis, meaning prop_down will be around 0.4. If you leave it at 0.5, 
  # this specific cell type will still be filtered out!
  is_inverse <- (percent_drop >= 0.05) & (prop_down >= 0.3) & (!is.na(plunge_rho)) & (plunge_rho <= -0.3)

  tibble(
    min_fit = min(y_pred, na.rm = TRUE),
    max_fit = max(y_pred, na.rm = TRUE),
    percent_drop = percent_drop,
    prop_down = prop_down,
    plunge_rho = plunge_rho,
    is_inverse = is_inverse
  )
}

# Apply to your data
all_trends_df <- celltype_day_raw %>%
  filter(n >= 100) %>%
  group_by(celltype_update) %>%
  nest() %>%
  mutate(trend_stats = map(data, evaluate_inverse_trend)) %>%
  unnest(trend_stats) %>%
  filter(celltype_update %in% celltype_filter$celltype_update) %>%
  filter(!is.na(is_inverse))

# View the convincing metrics
print(all_trends_df %>% 
        filter(is_inverse == TRUE) %>% 
        select(celltype_update, percent_drop, prop_down, plunge_rho))
```

```r
# 1. Generate the normalized curves (same as before)
get_spline_curve <- function(df) {
  if(nrow(df) < 10) return(NULL)
  fit <- lm(med_score_pred_glmm_Z5_0.7_0.4_n_6 ~ ns(med_prolif, df = 3), data = df)
  x_grid <- seq(min(df$med_prolif, na.rm=TRUE), max(df$med_prolif, na.rm=TRUE), length.out = 100)
  y_pred <- predict(fit, newdata = data.frame(med_prolif = x_grid))
  y_norm <- (y_pred - min(y_pred)) / (max(y_pred) - min(y_pred))
  tibble(Proliferation_Bin = 1:100, Normalized_Score = y_norm)
}

# --- NEW: Relabel strings immediately ---
heatmap_data <- celltype_day_raw %>%
  filter(n >= 100) %>%
  mutate(celltype_update = case_when(
    celltype_update == 'Definitive early erythroblasts (CD36-)' ~ 'Early erythroblasts',
    celltype_update == 'PMN myeloid-derived suppressor cells' ~ 'PMN-MDSCs',
    celltype_update == 'Pre-osteoblasts (Sp7+)' ~ 'Osteoblasts',
    celltype_update == 'Early chondrocytes' ~ 'Chondrocytes',
    celltype_update == 'Primitive erythroid cells' ~ 'Primitive erythroids',
    TRUE ~ celltype_update
  )) %>%
  group_by(celltype_update) %>%
  nest() %>%
  mutate(curves = map(data, get_spline_curve)) %>%
  unnest(curves) %>%
  filter(celltype_update %in% all_trends_df$celltype_update[all_trends_df$is_inverse == TRUE] | 
         celltype_update %in% c('Early erythroblasts', 'PMN-MDSCs', "Chondrocytes", "Osteoblasts",
                                "Primitive erythroids")) 

# 2. Reshape to wide format for clustering
wide_matrix <- heatmap_data %>%
  select(celltype_update, Proliferation_Bin, Normalized_Score) %>% 
  pivot_wider(names_from = Proliferation_Bin, values_from = Normalized_Score) %>%
  column_to_rownames("celltype_update") %>%
  as.matrix()

# 3. Perform Hierarchical Clustering
hc <- hclust(dist(wide_matrix))
clustered_order <- hc$labels[hc$order]

# Move the newly named Erythroblasts to the end
#target <- 'Early erythroblasts'
#clustered_order <- c(clustered_order[clustered_order != target], target)
clustered_order = c("Primitive erythroids", "Fibroblasts",  "Granular keratinocytes", 
                    "Osteoblasts", "Dorsal root ganglion neurons", 
                    "Chondrocytes", "PMN-MDSCs", "Myofibroblasts", 
                    "Early erythroblasts")


heatmap_data$celltype_update <- factor(heatmap_data$celltype_update, levels = clustered_order)

# 4. Plot
options(repr.plot.width=5, repr.plot.height=5, repr.plot.res=300)
ggplot(heatmap_data, aes(x = Proliferation_Bin, y = celltype_update, fill = Normalized_Score)) +
    geom_tile() +
    scale_fill_viridis_c(option = "inferno", direction = 1, name = "Scaled\nnucleolar\nscore",
                        guide = guide_colorbar(
                            direction = "horizontal",
                            barwidth = unit(4, "cm"),
                            barheight = unit(0.8, "cm"),
                            title.position = "left",
                            title.hjust = 1,
                            label.hjust = 0.5
                        )) + 
    # str_wrap is still useful for any other long names remaining
    scale_y_discrete(labels = function(x) stringr::str_wrap(x, width = 18)) +
    labs(x = "Proliferation Index (Low to High)", y = "") +
    theme_classic() +
    theme(axis.text.x = element_blank(), 
        axis.ticks.x = element_blank(),
        axis.line = element_line(color = "black"),
        axis.text = element_text(size = 13, color = "black"),
        axis.title = element_text(size = 13, color = "black"),
        legend.position = "top",
        legend.text = element_text(size = 13, color = "black"),
        legend.spacing.x = unit(0.4, 'cm'),
        legend.title = element_text(size = 13, margin = margin(r = 15))
    )
```

```r
cell_types_to_plot = all_trends_df$celltype_update[all_trends_df$is_inverse == TRUE]
text_size = 17

for (ctype in cell_types_to_plot) {
    ctype_data <- celltype_day_raw %>%
        filter(celltype_update == ctype, n >= 100) %>%
        arrange(day_numeric) %>%
        mutate(celltype_update = case_when(
                celltype_update == 'Definitive early erythroblasts (CD36-)' ~ 'Early erythroblasts',
                celltype_update == 'PMN myeloid-derived suppressor cells' ~ 'PMN-MDSCs',
                celltype_update == 'Pre-osteoblasts (Sp7+)' ~ 'Osteoblasts',
                celltype_update == 'Early chondrocytes' ~ 'Chondrocytes',
                celltype_update == 'Primitive erythroid cells' ~ 'Primitive erythroids',
            TRUE ~ celltype_update
        )) 
    
    if (ctype == 'Definitive early erythroblasts (CD36-)') ctype = 'Early erythroblasts'
    if (ctype == 'Early chondrocytes') ctype = 'Chondrocytes'
    if (ctype == 'PMN myeloid-derived suppressor cells') ctype = 'PMN-MDSCs'
    if (ctype == 'Pre-osteoblasts (Sp7+)') ctype = 'Osteoblasts'
    if (ctype == 'Primitive erythroid cells') ctype = 'Primitive erythroids'
    
    # --- 1. THE MATH WITH PADDING ---
    p_min <- min(ctype_data$med_prolif, na.rm=TRUE)
    p_max <- max(ctype_data$med_prolif, na.rm=TRUE)
    p_range <- p_max - p_min
    
    n_min <- min(ctype_data$med_score_pred_glmm_Z5_0.7_0.4_n_6, na.rm=TRUE)
    n_max <- max(ctype_data$med_score_pred_glmm_Z5_0.7_0.4_n_6, na.rm=TRUE)
    n_range <- n_max - n_min

    # ADJUST THIS: 0.7 means the red line takes up 70% of the black line's height
    # Increase to 0.9 for more zoom, decrease to 0.5 for less zoom
    target_fill <- 0.6 
    
    # Calculate multiplier and an offset to center the red line
    mult <- if(n_range == 0) 1 else (p_range * target_fill) / n_range
    p_center_offset <- (p_range * (1 - target_fill)) / 2

    options(repr.plot.width=4, repr.plot.height=3, repr.plot.res=300)
    g = ggplot(ctype_data, aes(x = day_numeric)) + 
        # Proliferation (Black)
        geom_point(aes(y = med_prolif), color = "gray60", alpha = 0.4, size = 1.2) +
        geom_smooth(aes(y = med_prolif, color = "Proliferation"), 
                    method = "lm", formula = y ~ ns(x, df = 5), se = FALSE, linewidth = 1.2) +

        # Nucleolar Score (Red) - Now with Padding and Centering
        geom_point(aes(y = (med_score_pred_glmm_Z5_0.7_0.4_n_6 - n_min) * mult + p_min + p_center_offset), 
                   color = "firebrick", alpha = 0.5, size = 1.2) +
        geom_smooth(aes(y = (med_score_pred_glmm_Z5_0.7_0.4_n_6 - n_min) * mult + p_min + p_center_offset, color = "Nuc Score"), 
                    method = "lm", formula = y ~ ns(x, df = 3), se = FALSE, linewidth = 1.2) +

        # --- 3. THE AXES ---
        scale_y_continuous(
            name = "Proliferation Index",
            # Primary axis rounding (optional)
            labels = scales::label_number(accuracy = 0.01), 

            sec.axis = sec_axis(
                ~ (. - p_min - p_center_offset) / mult + n_min, 
                name = "Nucleolar Score",
                # SECONDARY axis rounding to two digits
                labels = scales::label_number(accuracy = 0.01) 
            )
        ) +

        scale_color_manual(values = c("Proliferation" = "black", "Nuc Score" = "firebrick")) +
        labs(x = "Stage", title = ctype) +
        monocle3:::monocle_theme_opts() +
        theme(axis.line = element_line(color = "black"), 
              plot.title = element_text(color = "black", size = text_size, hjust = 0.5),
            axis.text.x = element_text(color = "black", size = text_size),
            legend.position = "none",
            legend.title = element_blank(),
            axis.text.y.left = element_text(color = "black", size = text_size),
            axis.title.y.left = element_text(color = "black", size = text_size),
            axis.title.y.right = element_text(color = "firebrick", size = text_size),
            axis.text.y.right = element_text(color = "firebrick", size = text_size),
            axis.title.x = element_text(color = "black", size = text_size)
        
        )
    
    print(g)
}
```

```r
celltype_behavior_analysis <- celltype_day_raw %>%
    filter(n >= 200) %>%
    group_by(celltype_update) %>%
    summarise(
        # 1. Global Correlation (The "Program")
        cor_global = cor(med_prolif, med_score_pred_glmm_Z5_0.7_0.4_n_6, method = "spearman"),
        
        # 2. Difference Correlation (The "Machinery")
        # We use diff() to look at day-to-day coupling
        cor_diff = cor(diff(med_prolif), diff(med_score_pred_glmm_Z5_0.7_0.4_n_6), method = "spearman"),
        
        n_days = n()
    ) %>%
    filter(n_days >= 15) %>% # Ensure enough points for two correlations
    mutate(coupling_gap = cor_diff - cor_global) %>%
    arrange(desc(coupling_gap))
```

```r
celltype_phase_flip <- celltype_day_raw %>%
    group_by(celltype_update) %>%
    summarise(
        # Calculate correlation for the first 5 available days
        early_cor = cor(head(med_prolif, 5), head(med_score_pred_glmm_Z5_0.7_0.4_n_6, 5), method = "spearman"),
        # Calculate correlation for the last 5 available days
        late_cor = cor(tail(med_prolif, 5), tail(med_score_pred_glmm_Z5_0.7_0.4_n_6, 5), method = "spearman")
    ) %>%
    mutate(flip_strength = abs(early_cor - late_cor)) %>%
    filter(sign(early_cor) != sign(late_cor)) # Only keep those that actually flipped sign

celltype_phase_flip
```

```r
library(dplyr)
library(purrr)


n_filter = 50
safe_cor_test <- possibly(cor.test, otherwise = NULL)

# 2. Create a helper function so we don't have to copy/paste the same code 3 times
calculate_tissue_cor <- function(data, target_col, score_name) {
    # Convert the column name string to a symbol so dplyr can use it
    target_sym <- sym(target_col)
    
    data %>%
        filter(!!target_sym > 0) %>%
        group_by(major_trajectory, celltype_update, embryo_id) %>%
        summarise(
            med_score = median(score_pred_glmm_Z5_0.7_0.4_n_6),
            med_target = median(!!target_sym),
            n = n(),
            .groups = "drop"
        ) %>%
        filter(n >= n_filter) %>%
        group_by(major_trajectory, celltype_update) %>%
        summarise(
            med_score = median(med_score),
            med_target = median(med_target),
            .groups = "drop"
        ) %>%
        group_by(major_trajectory) %>%
        filter(n() >= 5) %>% # Need at least 3 points per tissue
        tidyr::nest() %>%
        mutate(
            cor_model = map(data, ~ suppressWarnings(
                safe_cor_test(.x$med_score, .x$med_target, method = "spearman")
            ))
        ) %>%
        filter(!map_lgl(cor_model, is.null)) %>%
        mutate(
            score_type = score_name,
            spearman_rho = map_dbl(cor_model, "estimate"),
            p_value = map_dbl(cor_model, "p.value"),
            n_points = map_int(data, nrow)
        ) %>%
        select(score_type, major_trajectory, spearman_rho, p_value, n_points)
}

# 3. Run the function for all three scores
upr_tissue_cor     <- calculate_tissue_cor(df_all, "upr_score", "UPR")
#ups_tissue_cor     <- calculate_tissue_cor(metadata_remapped_filtered, "ups_score", "UPS Score")
granule_tissue_cor <- calculate_tissue_cor(df_all, "granule_score", "Stress granule")

# 4. Combine them all and sort
combined_tissue_cor <- bind_rows(upr_tissue_cor,granule_tissue_cor) %>%
    arrange(score_type, desc(spearman_rho))

# View the final output
print(combined_tissue_cor)
```

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res=300)

ggplot(combined_tissue_cor, aes(x = spearman_rho)) + 
    # The histogram bars
    geom_histogram(binwidth = 0.1, bins = 5, fill = "lightgray", color = "black", alpha = 0.8) +
    
    # The individual embryos at the base
    #geom_jitter(aes(y = 0.2), height = 0.15, size = 1.5, alpha = 0.7, color = "#2166ac") +
    
    # Reference line at 0
    geom_vline(xintercept = 0, linetype = "dashed", color = "red", linewidth = 0.8) +
    
    # Split the plot into rows based on the score type
    facet_grid(score_type ~ ., scales="free_y") +
    
    # Labels and theme
    labs(x = "Spearman Correlation (per Embryo)", y = "Count of Embryos") + 
    monocle3:::monocle_theme_opts() + theme(legend.position = "none") +
    theme(axis.text.x = element_text(color = "black", size = 12), 
          axis.text.y = element_text(color = "black", size = 12), 
          # Make the facet labels (row titles) look nice
          strip.text = element_text(color = "black", size = 10, face = "bold"),
          axis.ticks = element_line(color = "black"))

options(repr.plot.width=3, repr.plot.height=4, repr.plot.res=300)

ggplot(combined_tissue_cor, aes(x = score_type, y = spearman_rho, fill = score_type)) + 
    # 1. The Boxplot (outlier.shape = NA prevents duplicating points with the jitter)
    geom_boxplot(alpha = 0.7, color = "black", outlier.shape = NA, width = 0.6,
                 fatten = 0.75, lwd = 0.75) + 
    
    # 2. Overlay the individual tissue correlation points
    geom_jitter(width = 0.15, size = 1.5, alpha = 0.7, color = "black") + 
    
    # 3. Horizontal reference line at 0 (changed from geom_vline to geom_hline)
    geom_hline(yintercept = 0, linetype = "dashed", color = "red", linewidth = 0.8) +
    
    # 4. Labels and theme
    labs(x = "", y = "Spearman Correlation (per Tissue)") + 
    monocle3:::monocle_theme_opts() + 
    theme(legend.position = "none") +
    theme(axis.text.x = element_text(color = "black", size = 12, face = "bold"), 
          axis.text.y = element_text(color = "black", size = 12), 
          axis.ticks = element_line(color = "black"))
```
