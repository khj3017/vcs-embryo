```r
pseudobulk_cds = function(cds_sub, sample_group = "Cluster_group", cell_group = "Grouping") {
    cf_ccs = new_cell_count_set(cds_sub,
                            sample_group = sample_group,
                            cell_group = cell_group)
    
    pb_cds = hooke:::pseudobulk_ccs_for_states(cf_ccs)
    
    return(pb_cds)
    
}
```

```r
aggregate_gene_score = function(cds, gene_list=c("CCNB1", "CCNB2","CDK1"), name = "gene_list_score"){
  cds_gene_list_norm = cds[rowData(cds)$gene_short_name %in% gene_list,] 
  aggregate_marker_expression = counts(cds_gene_list_norm)
  aggregate_marker_expression = t(t(aggregate_marker_expression) / colData(cds_gene_list_norm)$Size_Factor)
  aggregate_marker_expression = Matrix::colSums(aggregate_marker_expression)
  aggregate_marker_expression = log(aggregate_marker_expression + 1)
  colData(cds)[[name]] = aggregate_marker_expression
  return(cds)
}

normalize <- function(x) {
    return((x- min(x)) /(max(x)-min(x)))
}

detect_genes_fixed = function (cds, min_expr = 0) {
    assertthat::assert_that(methods::is(cds, "cell_data_set"))
    assertthat::assert_that(is.numeric(min_expr))
    mat_bin <- counts(cds) > min_expr
    rowData(cds)$num_cells_expressed <- Matrix::rowSums(mat_bin)
    colData(cds)$num_genes_expressed <- Matrix::colSums(mat_bin)
    cds
}


estimate_cell_cycle = function(cds, g1s_markers, g2m_markers){
  cds_g1s = cds[rowData(cds)$gene_short_name %in% g1s_markers,] 
  aggregate_g1s_expression = counts(cds_g1s)
  aggregate_g1s_expression = t(t(aggregate_g1s_expression) / colData(cds_g1s)$Size_Factor)
  aggregate_g1s_expression = Matrix::colSums(aggregate_g1s_expression)
  
  cds_g2m = cds[rowData(cds)$gene_short_name %in% g2m_markers,] 
  aggregate_g2m_expression = counts(cds_g2m)
  aggregate_g2m_expression = t(t(aggregate_g2m_expression) / colData(cds_g2m)$Size_Factor)
  aggregate_g2m_expression = Matrix::colSums(aggregate_g2m_expression)
  #pData(cds)$cell_cycle_score = aggregate_g1s_expression
  cds$g1s_score = log(aggregate_g1s_expression+1)
  cds$g2m_score = log(aggregate_g2m_expression+1)
  cds$proliferation_index = log(aggregate_g1s_expression + aggregate_g2m_expression + 1)
  return(cds)
}
```

```r
res_df = NULL
for (traj in celltype_to_keep) {
    pb_cds_nuc_frac_sub = pb_cds_nuc_frac[, pb_cds_nuc_frac$trajectory_refined == traj]
    pb_cds_nuc_frac_sub = detect_genes_fixed(pb_cds_nuc_frac_sub, 0.01)

    rowdata = as.data.frame(rowData(pb_cds_nuc_frac_sub))

    genes = rowdata %>%
                filter(num_cells_expressed > 0.25*ncol(pb_cds_nuc_frac_sub))

    cm = normalized_counts(pb_cds_nuc_frac_sub)
    cm = cm[as.character(genes$gene_ID),]

    subject_id = as.factor(pb_cds_nuc_frac_sub$Embryo) 
    score_vec = pb_cds_nuc_frac_sub$avg_score

    # 3. Create a safe wrapper function for both Ranked (Spearman) and Standard (Pearson) rmcorr
    run_combined_rmcorr <- function(gene_expr, id, score) {

      # Dataframe for Pearson (raw continuous values)
      df_pearson <- data.frame(
        Embryo = id, 
        Score = score, 
        Expr = gene_expr
      )

      # Dataframe for Spearman (rank-transformed values)
      df_spearman <- data.frame(
        Embryo = id, 
        Score_Ranked = rank(score, ties.method = "average"), 
        Expr_Ranked = rank(gene_expr, ties.method = "average")
      )

      # Safely compute Repeated Measures Spearman
      spearman_res <- tryCatch({
        rmc_s <- rmcorr(participant = Embryo, measure1 = Score_Ranked, measure2 = Expr_Ranked, dataset = df_spearman)
        # Explicitly naming the outputs so they become clear column headers
        c(Spearman = rmc_s$r, Spearman_pvalue = rmc_s$p)
      }, error = function(e) c(Spearman = NA, Spearman_pvalue = NA))

      # Safely compute Repeated Measures Pearson
      pearson_res <- tryCatch({
        rmc_p <- rmcorr(participant = Embryo, measure1 = Score, measure2 = Expr, dataset = df_pearson)
        # Explicitly naming the outputs so they become clear column headers
        c(Pearson = rmc_p$r, Pearson_pvalue  = rmc_p$p)
      }, error = function(e) c(Pearson = NA, Pearson_pvalue = NA))

      # Return all four metrics combined
      return(c(spearman_res, pearson_res))
    }

    # 4. Apply the combined function across the genes
    rmcorr_combined_results <- apply(as.matrix(t(cm)), 2, run_combined_rmcorr, id = subject_id, score = score_vec)

    # 5. Format the results into your dataframe 
    gene_cor_df_tmp <- t(rmcorr_combined_results) %>%
      as.data.frame() %>%
      rownames_to_column("gene_ID") %>%
      left_join(., rowdata %>% dplyr::select(gene_ID, gene_short_name), by = "gene_ID") %>%
      mutate(trajectory_refined = traj,
             Spearman_FDR = p.adjust(Spearman_pvalue, method = "BH"),
             Pearson_FDR = p.adjust(Pearson_pvalue, method = "BH")) 
    
    if (is.null(res_df)) {
        res_df = gene_cor_df_tmp
    } else {
        res_df = rbind(res_df, gene_cor_df_tmp)
    }
    
    print(paste0(traj, ' Number of genes: ', nrow(genes)))
    print(pb_cds_nuc_frac_sub)
    flush.console()
     
}
```

```r
for (traj in c("Fibroblasts_Fap+")) {
    pb_cds_nuc_frac_sub = pb_cds_nuc_frac[, pb_cds_nuc_frac$trajectory_refined == traj]
    pb_cds_nuc_frac_sub = detect_genes_fixed(pb_cds_nuc_frac_sub, 0.01)

    rowdata = as.data.frame(rowData(pb_cds_nuc_frac_sub))

    genes = rowdata %>%
                filter(num_cells_expressed > 0.25*ncol(pb_cds_nuc_frac_sub))

    cm = normalized_counts(pb_cds_nuc_frac_sub)
    cm = cm[as.character(genes$gene_ID),]

    subject_id = as.factor(pb_cds_nuc_frac_sub$Embryo) 
    score_vec = pb_cds_nuc_frac_sub$avg_score

    # 3. Create a safe wrapper function for both Ranked (Spearman) and Standard (Pearson) rmcorr
    run_combined_rmcorr <- function(gene_expr, id, score) {

      # Dataframe for Pearson (raw continuous values)
      df_pearson <- data.frame(
        Embryo = id, 
        Score = score, 
        Expr = gene_expr
      )

      # Dataframe for Spearman (rank-transformed values)
      df_spearman <- data.frame(
        Embryo = id, 
        Score_Ranked = rank(score, ties.method = "average"), 
        Expr_Ranked = rank(gene_expr, ties.method = "average")
      )

      # Safely compute Repeated Measures Spearman
      spearman_res <- tryCatch({
        rmc_s <- rmcorr(participant = Embryo, measure1 = Score_Ranked, measure2 = Expr_Ranked, dataset = df_spearman)
        # Explicitly naming the outputs so they become clear column headers
        c(Spearman = rmc_s$r, Spearman_pvalue = rmc_s$p)
      }, error = function(e) c(Spearman = NA, Spearman_pvalue = NA))

      # Safely compute Repeated Measures Pearson
      pearson_res <- tryCatch({
        rmc_p <- rmcorr(participant = Embryo, measure1 = Score, measure2 = Expr, dataset = df_pearson)
        # Explicitly naming the outputs so they become clear column headers
        c(Pearson = rmc_p$r, Pearson_pvalue  = rmc_p$p)
      }, error = function(e) c(Pearson = NA, Pearson_pvalue = NA))

      # Return all four metrics combined
      return(c(spearman_res, pearson_res))
    }

    # 4. Apply the combined function across the genes
    rmcorr_combined_results <- apply(as.matrix(t(cm)), 2, run_combined_rmcorr, id = subject_id, score = score_vec)

    # 5. Format the results into your dataframe 
    gene_cor_df_tmp <- t(rmcorr_combined_results) %>%
      as.data.frame() %>%
      rownames_to_column("gene_ID") %>%
      left_join(., rowdata %>% dplyr::select(gene_ID, gene_short_name), by = "gene_ID") %>%
      mutate(trajectory_refined = traj,
             Spearman_FDR = p.adjust(Spearman_pvalue, method = "BH"),
             Pearson_FDR = p.adjust(Pearson_pvalue, method = "BH")) 
    
    if (is.null(res_df)) {
        res_df = gene_cor_df_tmp
    } else {
        res_df = rbind(res_df, gene_cor_df_tmp)
    }
    
    print(paste0(traj, ' Number of genes: ', nrow(genes)))
    print(pb_cds_nuc_frac_sub)
    flush.console()
     
}
```

```r
cm = normalized_counts(pb_cds_nuc_frac_all)

# 2. Extract vectors for grouping and scoring
# Assuming 'Embryo' is the column denoting your individual subjects
subject_id = as.factor(pb_cds_nuc_frac_all$Embryo) 
score_vec = pb_cds_nuc_frac_all$avg_score

# 3. Create a combined wrapper to calculate both RM correlations and p-values simultaneously
run_combined_rmcorr <- function(gene_expr, id, score) {
    # Standard dataframe for Pearson
    df_pearson <- data.frame(Embryo = id, Score = score, Expr = gene_expr)

    # Rank-transformed dataframe for Spearman
    df_spearman <- data.frame(
        Embryo = id,
        Score_Ranked = rank(score, ties.method = "average"),
        Expr_Ranked = rank(gene_expr, ties.method = "average")
    )

    # Calculate Repeated Measures Spearman
    spearman_res <- tryCatch({
        rmc_s <- rmcorr(participant = Embryo, measure1 = Score_Ranked, measure2 = Expr_Ranked, dataset = df_spearman)
        c(Spearman = rmc_s$r, Spearman_pvalue = rmc_s$p)
    }, error = function(e) c(Spearman = NA, Spearman_pvalue = NA))

    # Calculate Repeated Measures Pearson
    pearson_res <- tryCatch({
        rmc_p <- rmcorr(participant = Embryo, measure1 = Score, measure2 = Expr, dataset = df_pearson)
        c(Pearson = rmc_p$r, Pearson_pvalue = rmc_p$p)
    }, error = function(e) c(Pearson = NA, Pearson_pvalue = NA))

    # Return all metrics combined in specific order
    return(c(spearman_res, pearson_res))
}

# 4. Apply the function across all genes in the count matrix
rmcorr_results = apply(as.matrix(t(cm)), 2, run_combined_rmcorr, id = subject_id, score = score_vec)

# 5. Format to match your exact output structure
gene_cor_df_all = t(rmcorr_results) %>%
    as.data.frame() %>%
    rownames_to_column("gene_ID") %>%
    left_join(., rowdata %>% dplyr::select(gene_ID, gene_short_name), by = "gene_ID") %>%
    na.omit()
                            
gene_cor_df_all <- gene_cor_df_all %>%
  mutate(
    # Create new columns for the FDR-adjusted p-values
    Spearman_FDR = p.adjust(Spearman_pvalue, method = "BH"),
    Pearson_FDR = p.adjust(Pearson_pvalue, method = "BH")
  ) 
```

```r
dynamic_lineages <- c("GABAergic", "Glutamatergic", "Mesoderm")
                      #"White_blood_cells", "Epithelium")

# 2. Run Stouffer's Method
stouffer_df <- res_df %>%
  # Keep only dynamic lineages
  filter(!trajectory_refined %in% dynamic_lineages) %>%
  
  # Protect against p-values of exactly 0 (which cause infinite Z-scores in R)
  mutate(Spearman_pvalue = ifelse(Spearman_pvalue == 0, 1e-300, Spearman_pvalue)) %>%
  
  # Step A: Convert two-tailed p-values to directional Z-scores
  mutate(
    # qnorm converts p-value to Z-score. 
    # We multiply by the sign() of the Spearman R to keep the direction!
    Z_score = sign(Spearman) * qnorm(Spearman_pvalue / 2, lower.tail = FALSE)
  ) %>%
  
  # Step B: Group by gene to combine across trajectories
  group_by(gene_ID, gene_short_name) %>%
  summarize(
    mean_Spearman = mean(Spearman, na.rm = TRUE),
    median_Spearman = median(Spearman, na.rm = TRUE),
    k_trajectories = n(),                             # How many lineages was this gene tested in?
    sum_Z = sum(Z_score, na.rm = TRUE),               # Add the Z-scores together
    n_positive_0.2 = sum(Spearman > 0.2, na.rm = TRUE), 
    n_positive_0.3 = sum(Spearman > 0.3, na.rm = TRUE), 
    n_positive_0.4 = sum(Spearman > 0.4, na.rm = TRUE), 
    n_negative = sum(Spearman < -0.3, na.rm = TRUE),     # Count negative votes
    .groups = "drop"
  ) %>%
  
  # Step C: Calculate Stouffer's Z and convert back to a Meta P-value
  mutate(
    Stouffer_Z = sum_Z / sqrt(k_trajectories),        # The core Stouffer formula
    Meta_Pvalue = 2 * pnorm(-abs(Stouffer_Z)),        # Convert combined Z back to a two-tailed p-value
    Meta_FDR = p.adjust(Meta_Pvalue, method = "BH")   # Finally, apply the FDR penalty to the meta p-value!
  ) %>%
  
  # Step D: Sort by the strongest consistent positive correlations
  arrange(desc(Stouffer_Z))
```

```r
filter_intra_inter_genes = function(res_df, gene_cor_df, intra_corr = 0.3, inter_corr = 0.7, 
                                    fdr_cutoff = 0.05, n_lineage = 5, negative_corr = FALSE) {
    

    m = ifelse(negative_corr, -1, 1)

    genes_test = res_df %>%
        filter(Spearman_FDR < fdr_cutoff & Pearson_FDR < fdr_cutoff) %>%
        #filter(Spearman_pvalue < fdr_cutoff & Pearson_pvalue < fdr_cutoff) %>%
        filter((Spearman * m) > intra_corr & (Pearson * m) > (intra_corr-0.1)) %>%
        group_by(gene_ID, gene_short_name) %>% 
        count() %>%
        arrange(-n) %>%
        filter(n >= n_lineage) %>%
        left_join(., gene_cor_df %>%
                        filter(Spearman_FDR < fdr_cutoff & Pearson_FDR < fdr_cutoff) %>%
                        #filter(Spearman_pvalue < fdr_cutoff & Pearson_pvalue < fdr_cutoff) %>%
                          select(gene_ID, gene_short_name, Spearman, Pearson),
                  by = c("gene_ID", "gene_short_name")) %>%

        filter((Spearman * m) > inter_corr & (Pearson * m) > (inter_corr-0.1)) %>% 
        as.data.frame() 

    filt_genes = res_df %>%
        filter(gene_short_name %in% genes_test$gene_short_name) %>%
            #filter(Spearman_pvalue < fdr_cutoff & (Spearman * -m) > 0.4 &
            #       Pearson_pvalue < fdr_cutoff & (Pearson * -m) > 0.4)
         filter(Spearman_FDR < fdr_cutoff & (Spearman * -m) > 0.4 &
                Pearson_FDR < fdr_cutoff & (Pearson * -m) > 0.4)

    genes_test_filt = genes_test %>%
                    filter(!gene_short_name %in% filt_genes$gene_short_name)


    if (negative_corr) {
        genes_test_filt = genes_test_filt %>% arrange(Spearman)
    } else {
        genes_test_filt = genes_test_filt %>% arrange(-Spearman)
    }
    
    res = list()
    res[[1]] = genes_test
    res[[2]] = genes_test_filt
    
    return(res)
}
```

```r
print_gene_comparison <- function(orig_df, other_df, label = "Threshold Set") {
  # 1. Extract and clean names
  orig_names  <- sort(unique(orig_df$gene_short_name))
  other_names <- sort(unique(other_df$gene_short_name))
  
  # 2. Calculate differences
  only_orig  <- setdiff(orig_names, other_names)
  only_other <- setdiff(other_names, orig_names)
  shared     <- intersect(orig_names, other_names)
  
  # 3. Nicely print the summary
  cat("\n", paste0(rep("=", 40), collapse = ""), "\n")
  cat(" COMPARISON: Original vs.", label, "\n")
  cat(paste0(rep("-", 40), collapse = ""), "\n")
  cat("Shared Genes: ", length(shared), "\n")
  cat(paste(shared, collapse = ", "), "\n")
  
  cat("\n[ ONLY IN ORIGINAL ] (Count:", length(only_orig), ")\n")
  if(length(only_orig) > 0) {
    cat(paste(only_orig, collapse = ", "), "\n")
  } else {
    cat("None\n")
  }
  
  cat("\n[ ONLY IN", toupper(label), "] (Count:", length(only_other), ")\n")
  if(length(only_other) > 0) {
    cat(paste(only_other, collapse = ", "), "\n")
  } else {
    cat("None\n")
  }
  cat(paste0(rep("=", 40), collapse = ""), "\n\n")
}
```

```r
library(patchwork)

# 1. Update the function to optionally hide the x-axis
plot_gene_score <- function(gene_name, show_xaxis = FALSE) {
  p <- as.data.frame(colData(pb_cds_nuc_frac_all)) %>%
    ggplot(., aes(x = avg_score, y = .data[[gene_name]])) +
    geom_point(size = 0.4, alpha = 0.6) + 
    geom_smooth(method = "lm", formula = y ~ poly(x, 3), color = "2", se = T) +
    labs(x = "Nucleolar fraction score", 
         y = paste0(gene_name, " expression")) +  
    monocle3:::monocle_theme_opts() + 
    theme(
      axis.text.y = element_text(color = "black", size = 11),
      axis.title.y = element_text(size = 12),
      legend.position = "none",
      axis.ticks = element_line(color = "black"),
      # Reduce plot margins to "squish" the plots together
      plot.margin = margin(t = 5, r = 5, b = 5, l = 5) 
    )

  if (!show_xaxis) {
    p <- p + theme(
      axis.title.x = element_blank(),
      axis.text.x = element_blank(),
      axis.ticks.x = element_blank()
    )
  } else {
    p <- p + theme(
      axis.text.x = element_text(color = "black", size = 11),
      axis.title.x = element_text(size = 12)
    )
  }
  return(p)
}

# 2. Generate plots (only show_xaxis for the last one)
p1 <- plot_gene_score("Mthfd1l", show_xaxis = FALSE)
#p2 <- plot_gene_score("Alkbh1", show_xaxis = FALSE)
p3 <- plot_gene_score("Ubap2", show_xaxis = TRUE)

# 3. Combine and "Squish"
options(repr.plot.width = 5, repr.plot.height = 5, repr.plot.res = 300)

# Use & theme(plot.margin = ...) to globally reduce whitespace between panels
(p1 /  p3) + 
  plot_layout(guides = "collect") & 
  theme(plot.margin = margin(t = 0, b = 0, r = 5, l = 5))
```

```r
# 2. Define Trajectory Order
traj_order = c("CNS_neurons", "Intermediate_progenitors", "Neuroectoderm_and_glia",
               "Epithelial", "Mesenchymal", "Myogenic", 
               "Myeloid", "Hepatic", "Erythroid")

celltype_to_keep = c("Mesoderm", "Muscle_cells", "Epithelium",
               "White_blood_cells", "Hepatocytes", "Definitive_erythroid")

metadata_plot_base = as.data.frame(colData(pb_cds_nuc_frac_all)) %>%
    filter(trajectory_refined %in% celltype_to_keep)


# 3. Clean and Factorize Metadata once
meta_df_ordered <- metadata_plot_base %>%
    filter(!trajectory_refined %in% c("CNS_neurons", "Intermediate_progenitors")) %>%
    mutate(trajectory_refined = case_match(trajectory_refined,
                                           "White_blood_cells" ~ "Myeloid",
                                           "Muscle_cells" ~ "Myogenic",
                                           "Epithelium" ~ "Epithelial",
                                           "Mesoderm" ~ "Mesenchymal",
                                           "Hepatocytes" ~ "Hepatic",
                                           "Definitive_erythroid" ~ "Erythroid",
                                            .default = trajectory_refined),
           trajectory_refined = gsub('_', ' ', trajectory_refined),
           trajectory_refined = factor(trajectory_refined, levels = gsub('_', ' ', traj_order))) %>%
    group_by(trajectory_refined) %>%
    arrange(avg_score, .by_group = TRUE) %>%
    mutate(local_score_gradient = (avg_score - min(avg_score)) / (max(avg_score) - min(avg_score))) %>%
    ungroup()

# Get shared resources
cell_ids <- meta_df_ordered$pseudobulk_id
cm <- normalized_counts(pb_cds_nuc_frac_all)
active_trajs <- levels(meta_df_ordered$trajectory_refined)

# Define Colors
score_col_fun = colorRamp2(c(0, 1), c("#f0f0f0", "#555555"))
z_col_fun = colorRamp2(c(-2, 0, 2), c("#2166AC", "white", "#B2182B"))


my_new_label <- "Nucleolar fraction"

top_anno <- HeatmapAnnotation(
    # 1. Use a simple internal name for mapping
    nuc_frac = meta_df_ordered$local_score_gradient,

    # 2. Assign the color function to that internal name
    col = list(nuc_frac = score_col_fun),
    
    # 3. Define the actual DISPLAY label with \n
    annotation_label = " ", 
    #annotation_name_side = "left", 
    annotation_name_rot = 0,
    
    # 4. Fine-tune the styling and vertical spacing
    annotation_name_gp = gpar(
        fontface = "bold", 
        fontsize = 13,
        # Adjust lineheight if the gap between "Nucleolar" and "fraction" is too wide
        lineheight = 0.9 
    ),
    
    # 5. Ensure the bar itself is tall enough to match the two lines of text
    height = unit(0.6, "cm"),
    
    show_annotation_name = TRUE,
    
    annotation_legend_param = list(
        nuc_frac = list(
            title = "Scaled\nnucleolar\nfraction", 
            direction = "vertical", 
            title_position = "topleft"
        )
    )
)



for (set_name in names(stouffer_df_list)) {
    
    # 1. DATA SAFETY CHECK
    # This prevents the loop from "stopping" if one set has bad data
    curr_genes_df <- stouffer_df_list[[set_name]]
    valid_genes <- intersect(curr_genes_df$gene_ID, rownames(cm))
    
    if (length(valid_genes) < 2) {
        cat("\n--- Skipping", set_name, ": Not enough matching genes ---\n")
        next
    }
    
    # 2. FORCE JUPYTER TO START A NEW PLOT AREA
    # This is the secret for Jupyter: Resetting the repr inside the loop
    options(repr.plot.width=11.5, repr.plot.height=12, repr.plot.res=300)
    grid.newpage() 
    
    # 3. MATRIX PROCESSING
    heatmap_matrix <- cm[valid_genes, cell_ids]
    
    local_scaled_list <- list()
    for (traj in active_trajs) {
        lineage_cells <- meta_df_ordered$pseudobulk_id[meta_df_ordered$trajectory_refined == traj]
        if(length(lineage_cells) == 0) next
        
        mat_subset <- heatmap_matrix[, lineage_cells, drop = FALSE]
        scaled_subset <- t(scale(t(mat_subset)))
        scaled_subset[is.na(scaled_subset)] <- 0 
        local_scaled_list[[as.character(traj)]] <- scaled_subset
    }
    
    heatmap_matrix_local_scaled <- do.call(cbind, local_scaled_list)
    heatmap_matrix_local_scaled <- heatmap_matrix_local_scaled[, cell_ids]
    
    # Update row names for this specific set
    gene_labels <- curr_genes_df$gene_short_name[match(rownames(heatmap_matrix_local_scaled), curr_genes_df$gene_ID)]
    rownames(heatmap_matrix_local_scaled) <- gene_labels

    # 4. HEATMAP DEFINITION
    ht <- Heatmap(
        heatmap_matrix_local_scaled,            
        col = z_col_fun,              
        name = "Scaled\nexpression",
        #column_title = paste("Set:", set_name), 
        top_annotation = top_anno,
        column_split = meta_df_ordered$trajectory_refined, 
        column_gap = unit(3, "mm"),
        cluster_columns = FALSE,                
        cluster_rows = TRUE,                    
        show_column_names = FALSE,
        row_names_gp = gpar(fontsize = 13), 
        column_title_gp = gpar(fontsize = 13)
    )
    
    # 5. THE EXPLICIT DRAW (Wrapped in print)
    # Inside a loop, draw() MUST be wrapped in print() for Jupyter/IRkernel
    cat("\nRendering Heatmap for:", set_name, "...\n")
    print(draw(ht, 
               heatmap_legend_side = "right", 
               annotation_legend_side = "right", 
               merge_legend = TRUE))
}
```

```r
gene_stats <- data.frame(
  gene = rownames(cm), # Replace with your gene names
  rho = gene_cor_df$Spearman
)

# 2. Process the data to find the knee
knee_data <- gene_stats %>%
  mutate(abs_rho = abs(rho)) %>%
  arrange(desc(abs_rho)) %>%
  mutate(rank = row_number())

# 3. Mathematical calculation of the "Knee" (max distance from the diagonal)
# This draws an imaginary line from your highest correlation to your lowest,
# and finds the point on your curve that is furthest from that line.
x <- knee_data$rank
y <- knee_data$abs_rho
m <- (y[length(y)] - y[1]) / (x[length(x)] - x[1])
b <- y[1] - m * x[1]
distances <- abs(m * x - y + b) / sqrt(m^2 + 1)
knee_index <- which.max(distances)
knee_threshold <- knee_data$abs_rho[knee_index]

print(paste("Mathematical Knee Threshold is at absolute rho >", round(knee_threshold, 3)))

# 4. Plot the curve with the calculated threshold
ggplot(knee_data, aes(x = rank, y = abs_rho)) +
  geom_line(size = 1, color = "black") +
  geom_hline(yintercept = knee_threshold, color = "red", linetype = "dashed", size = 1) +
  geom_vline(xintercept = knee_index, color = "blue", linetype = "dotted", size = 1) +
  labs(title = "Rank-Ordered Correlation Plot",
       x = "Gene Rank",
       y = "Absolute Spearman Correlation (|ρ|)") +
  annotate("text", x = max(knee_data$rank) * 0.8, y = knee_threshold + 0.05, 
           label = paste("Cutoff: |ρ| >", round(knee_threshold, 3)), color = "red") +
  theme_classic() +
  theme(plot.title = element_text(hjust = 0.5, face = "bold"))
```

```r
pb_cds_nuc_frac_all = detect_genes_fixed(pb_cds_nuc_frac_all, 0.01)
rowdata = as.data.frame(rowData(pb_cds_nuc_frac_all))
gene_IDs = rowdata %>% filter(num_cells_expressed > 0.25 * ncol(pb_cds_nuc_frac_all)) %>%
                rownames()
```

```r
# Note: This requires your matrix of expression and your vector of cell scores.
# expression_matrix: genes as rows, cells as columns
# actual_scores: a numeric vector of your avg_score, in the exact same order as matrix columns

n_permutations <- 100 
null_correlations <- numeric()

print("Running permutations... this may take a minute.")

for (i in 1:n_permutations) {
  # 1. Randomly scramble the scores across the cells
  shuffled_scores <- sample(pb_cds_nuc_frac_all$avg_score)
  
  # 2. Calculate Spearman correlation of all genes against the FAKE scores
  # Use use="pairwise.complete.obs" to handle any NAs safely
  perm_rhos <- apply(cm[gene_IDs, ], 1, function(gene_expr) {
    cor(gene_expr, shuffled_scores, method = "spearman", use = "pairwise.complete.obs")
  })
  
  # 3. Store the absolute maximum correlation generated purely by chance in this run
  # We use the max() to strictly control the Family-Wise Error Rate (FWER)
  null_correlations <- c(null_correlations, max(abs(perm_rhos), na.rm = TRUE))
}

# 4. Calculate the rigorous p-value thresholds based on the null distribution
# The 99th percentile means only 1% of random shuffles ever achieved a correlation this high
empirical_threshold_01 <- quantile(null_correlations, 0.99)
empirical_threshold_05 <- quantile(null_correlations, 0.95)

print(paste("To achieve an Empirical FDR < 0.01, use absolute rho >", round(empirical_threshold_01, 3)))
print(paste("To achieve an Empirical FDR < 0.05, use absolute rho >", round(empirical_threshold_05, 3)))

# 5. Visualize the Null Distribution to prove it to the reviewers
hist(null_correlations, breaks = 20, col = "lightgray", border = "white",
     main = "Null Distribution of Spearman Correlations\n(100 Permutations)",
     xlab = "Maximum Absolute Spearman |ρ| Expected by Chance",
     ylab = "Frequency")
abline(v = empirical_threshold_01, col = "red", lwd = 2, lty = 2)
legend("topright", legend = paste("Empirical FDR 0.01\n(|ρ| >", round(empirical_threshold_01, 3), ")"), 
       col = "red", lty = 2, lwd = 2, bty = "n")
```

```r
n_permutations <- 100 

# We will store the 99th percentile threshold for each lineage in this list
lineage_thresholds <- list()

print("Running lineage-specific gene filtering and permutations...")

for (traj in unique(meta_df$trajectory_refined)) {
  print(paste("Processing lineage:", traj))
  
  # 1. Isolate the CDS and cells for this specific lineage
  pb_cds_nuc_frac_sub <- pb_cds_nuc_frac_all[, pb_cds_nuc_frac_all$trajectory_refined == traj]
  pb_cds_nuc_frac_sub <- detect_genes_fixed(pb_cds_nuc_frac_sub, 0.01)
  
  # 2. Filter genes dynamically based on expression in > 25% of cells in THIS lineage
  rowdata <- as.data.frame(rowData(pb_cds_nuc_frac_sub))
  genes <- rowdata %>% filter(num_cells_expressed > 0.25 * ncol(pb_cds_nuc_frac_sub))
  
  # 3. Extract and uncompress only the filtered expression matrix
  cm <- normalized_counts(pb_cds_nuc_frac_sub)
  cm <- cm[as.character(genes$gene_ID), ]
  dense_mat_subset <- as.matrix(t(cm))
  
  # Extract actual scores for this lineage
  cells_in_lineage <- colnames(pb_cds_nuc_frac_sub)
  scores_subset <- meta_df$avg_score[meta_df$pseudobulk_id %in% cells_in_lineage]
  
  # 4. Run the 100 permutations strictly for this lineage
  null_correlations_this_lineage <- numeric(n_permutations)
  
  for (i in 1:n_permutations) {
    # Shuffle the scores
    shuffled_scores <- sample(scores_subset)
    
    # Calculate Spearman correlation
    # suppressWarnings hides the harmless divide-by-zero warnings for zero-variance genes
    perm_rhos <- suppressWarnings(
      cor(dense_mat_subset, shuffled_scores, method = "spearman", use = "pairwise.complete.obs")
    )
    
    # Store the max correlation generated by chance for this run
    null_correlations_this_lineage[i] <- max(abs(perm_rhos), na.rm = TRUE)
  }
  
  # 5. Calculate and save the rigorous intra-lineage threshold for this specific lineage
  traj_threshold <- quantile(null_correlations_this_lineage, 0.99)
  lineage_thresholds[[traj]] <- traj_threshold
  print(paste("  -> FDR < 0.01 threshold for", traj, ": |rho| >", round(traj_threshold, 3)))
}

# 6. Find the absolute strictest threshold across all lineages to use as a global cutoff
strictest_global_threshold <- max(unlist(lineage_thresholds))
print(paste("=================================================================="))
print(paste("STRICTEST GLOBAL CUTOFF (Use this for the paper): |rho| >", round(strictest_global_threshold, 3)))
print(paste("=================================================================="))
```

```r
for (set_name in names(stouffer_df_list)) {
    
    cat("\n--- Processing Model for:", set_name, "---\n")
    
    # 1. PREPARE DATA
    # Use the specific column created for this set
    score_col <- paste0("gene_score_", set_name)
    
    metadata_fit = as.data.frame(colData(pb_cds_nuc_frac_all))
    
    # Ensure our x-axis variable is normalized for the model
    metadata_fit$norm_x = normalize(metadata_fit[[score_col]])
    
    # Squeeze the response variable (avg_score) to fit strictly within (0, 1)
    n <- nrow(metadata_fit)
    metadata_fit$fraction_squeezed <- (normalize(metadata_fit$avg_score) * (n - 1) + 0.5) / n

    # 2. BUILD THE MIXED-EFFECTS MODEL
    # Random effect (1 | Embryo) accounts for batch variation across replicates
    model_glmm <- glmmTMB(
      fraction_squeezed ~ norm_x + (1 | Embryo), 
      data = metadata_fit, 
      family = beta_family(link = "cloglog")
    )

    # 3. PREDICT RESPONSE
    # Generate predictions for the fixed effect across the 0-1 range
    pred_universal <- predict_response(
        model_glmm, 
        terms = "norm_x [0:1, by=0.01]",
        type = "fixed"
    )

    # 4. PLOT
    options(repr.plot.width=5, repr.plot.height=3.5, repr.plot.res=300)
    
    p <- ggplot() +
      # Points colored by Embryo to visualize the random effect spread
      geom_point(data = metadata_fit, 
                 aes(x = norm_x, y = fraction_squeezed, color = Embryo), 
                 alpha = 0.4, size = 0.5) +
      
      # Fixed Effect Regression Line
      geom_line(data = pred_universal, 
                aes(x = x, y = predicted), 
                color = "red", linewidth = 0.8) +
      
      # 95% CI Ribbon
      geom_ribbon(data = pred_universal, 
                  aes(x = x, ymin = conf.low, ymax = conf.high), 
                  fill = "red", alpha = 0.15) +
      
      labs(title = paste("Model:", set_name),
           x = "Normalized Gene Set Score", 
           y = "Squeezed Nucleolar Fraction",
           color = "Embryo") +
      monocle3:::monocle_theme_opts() +
      theme(legend.position = "right", aspect.ratio = 1)

    print(p)

    # 5. SAVE RESULTS
    # Using the set_name in the filename to prevent overwriting
    model_filename <- paste0("nucleolar_model_glmm_", set_name, ".RDS")
    gene_filename <- paste0("nucleolar_genes_", set_name, ".RDS")
    
    saveRDS(model_glmm, model_filename)
    saveRDS(stouffer_df_list[[set_name]], gene_filename)
    
    cat("Done. Model saved as:", model_filename, "\n")
}
```

```r
rm_data <- metadata_fit %>%
    mutate(trajectory_refined = case_match(trajectory_refined,
                                           "White_blood_cells" ~ "Myeloid",
                                           "Muscle_cells" ~ "Myogenic",
                                           "Epithelium" ~ "Epithelial",
                                           "Fibroblasts_Fap+" ~ "Fibroblasts",
                                           "Hepatocytes" ~ "Hepatic",
                                           "Definitive_erythroid" ~ "Erythroid",
                                           "Upper_layer_neurons" ~ "Upper layer\nneurons",
                                           "GABAergic" ~ "GABAergic\nneurons",
                                           "Glutamatergic" ~ "Glutamatergic\nneurons",
                                           .default = trajectory_refined)) %>%
    # Ensure Embryo is a factor for rmcorr
    mutate(Embryo = factor(Embryo))

rm_ranked_results <- rm_data %>%
    # Remove rows with NA scores
    filter(!is.na(avg_score), !is.na(gene_targets)) %>%
    group_by(trajectory_refined) %>%
    group_modify(~ {
        # A. Filter for embryos that have AT LEAST 3 cells
        valid_embryos <- .x %>%
            count(Embryo) %>%
            filter(n >= 3) %>%
            pull(Embryo)
        
        df_sub <- .x %>% filter(Embryo %in% valid_embryos)
        
        # B. Check if we have at least 2 valid embryos left
        if (length(unique(df_sub$Embryo)) < 2) {
            return(data.frame(r_rm = NA, p_val = NA, n_embryos = length(unique(df_sub$Embryo))))
        }
        
        # C. Run Ranked rmcorr
        tryCatch({
            # Ranking within the trajectory group
            res <- rmcorr(participant = Embryo, 
                          measure1 = rank(avg_score), 
                          measure2 = rank(gene_targets), 
                          dataset = df_sub)
            
            data.frame(
                r_rm = round(res$r, 3),
                p_val = res$p,
                n_embryos = length(unique(df_sub$Embryo))
            )
        }, error = function(e) {
            data.frame(r_rm = NA, p_val = NA, n_embryos = length(unique(df_sub$Embryo)))
        })
    }) %>%
    ungroup()

rm_ranked_results <- rm_ranked_results %>%
    mutate(label = ifelse(is.na(r_rm), 
                          'Insuff. Data', # Valid expression for parse=TRUE
                          paste0("ρ = ", r_rm)))
        
# 3. View the results
print(rm_ranked_results)
```

```r
geneset_df_filtered <- gene_set_BP_df %>%
    group_by(gs_name) %>%
    mutate(n_genes = n()) %>%
    filter(n_genes >= 15 & n_genes <= 500) %>%
    select(-n_genes) %>%
    ungroup()

# Check how many sets survived
length(unique(geneset_df_filtered$gs_name))

# Convert geneset dataframe to a named list (standard format for pathways)
gs_list <- split(geneset_df_filtered$gene_short_name, geneset_df_filtered$gs_name)

# Function to remove highly overlapping sets (Greedy approach)
reduce_redundancy <- function(pathway_list, max_overlap = 0.85) {
    # Sort by size (keep the smaller, more specific pathways when there is a tie)
    pathway_list <- pathway_list[order(sapply(pathway_list, length))]
    
    keep_sets <- c()
    
    for (i in seq_along(pathway_list)) {
        current_set <- pathway_list[[i]]
        current_name <- names(pathway_list)[i]
        
        if (length(keep_sets) == 0) {
            keep_sets <- c(keep_sets, current_name)
            next
        }
        
        # Calculate max overlap with already kept sets
        # Overlap = (intersection) / (size of current set)
        overlaps <- sapply(keep_sets, function(kept_name) {
            kept_set <- pathway_list[[kept_name]]
            length(intersect(current_set, kept_set)) / length(current_set)
        })
        
        # If the current set is less than 85% redundant with what we've kept, keep it
        if (max(overlaps) < max_overlap) {
            keep_sets <- c(keep_sets, current_name)
        }
    }
    return(keep_sets)
}

# Run the reducer
non_redundant_names <- reduce_redundancy(gs_list, max_overlap = 0.85)

# Final clean dataframe
geneset_df_clean <- geneset_df_filtered %>%
    filter(gs_name %in% non_redundant_names)

cat("Reduced from 7600 to:", length(non_redundant_names), "sets\n")
```

```r
run_combined_rmcorr <- function(gene_expr, id, score) {
  df_pearson <- data.frame(Embryo = id, Score = score, Expr = gene_expr)
  df_spearman <- data.frame(Embryo = id, 
                            Score_Ranked = rank(score, ties.method = "average"), 
                            Expr_Ranked = rank(gene_expr, ties.method = "average"))
  
  spearman_res <- tryCatch({
    rmc_s <- rmcorr(participant = Embryo, measure1 = Score_Ranked, measure2 = Expr_Ranked, dataset = df_spearman)
    c(Spearman = rmc_s$r, Spearman_pvalue = rmc_s$p)
  }, error = function(e) c(Spearman = NA, Spearman_pvalue = NA))
  
  pearson_res <- tryCatch({
    rmc_p <- rmcorr(participant = Embryo, measure1 = Score, measure2 = Expr, dataset = df_pearson)
    c(Pearson = rmc_p$r, Pearson_pvalue  = rmc_p$p)
  }, error = function(e) c(Pearson = NA, Pearson_pvalue = NA))
  
  return(c(spearman_res, pearson_res))
}
```

```r
# 3. Run the lineage loop
res_df_gs <- NULL

for (traj in unique(pb_cds_nuc_frac$trajectory_refined)) {
  
  # Extract just the metadata for this trajectory
  # No need to touch the large sparse matrices anymore!
  meta_sub <- as.data.frame(colData(pb_cds_nuc_frac)) %>%
    filter(trajectory_refined == traj)
  
  subject_id <- as.factor(meta_sub$Embryo) 
  score_vec <- meta_sub$avg_score
  
  # Safety check for sufficient subjects/cells
  if (length(unique(subject_id)) < 2 || nrow(meta_sub) < 3) {
    cat("\n--- Skipping", traj, ": Not enough embryos/cells ---\n")
    next
  }
  
  traj_res_list <- list()
  
  # 4. Iterate through each geneset column in the metadata
  for (gs in unique_genesets) {
    gs_expr <- meta_sub[[gs]]
    
    # Skip if the aggregate expression is 0 across all cells in this trajectory 
    # (prevents rmcorr zero-variance failure)
    if (sum(gs_expr, na.rm = TRUE) == 0) {
      traj_res_list[[gs]] <- c(Spearman = NA, Spearman_pvalue = NA, Pearson = NA, Pearson_pvalue = NA)
      next
    }
    
    res <- run_combined_rmcorr(gene_expr = gs_expr, id = subject_id, score = score_vec)
    traj_res_list[[gs]] <- res
  }
  
  # 5. Format results into a dataframe
  gs_cor_df_tmp <- do.call(rbind, traj_res_list) %>%
    as.data.frame() %>%
    rownames_to_column("gs_name") %>%
    mutate(trajectory_refined = traj,
           Spearman_FDR = p.adjust(Spearman_pvalue, method = "BH"),
           Pearson_FDR = p.adjust(Pearson_pvalue, method = "BH")) 
  
  # Bind to main results dataframe
  if (is.null(res_df_gs)) {
    res_df_gs <- gs_cor_df_tmp
  } else {
    res_df_gs <- rbind(res_df_gs, gs_cor_df_tmp)
  }
  
  print(paste0(traj, ': Processed ', length(unique_genesets), ' genesets'))
  flush.console()
}
```

```r
dynamic_lineages <- c("GABAergic", "Glutamatergic", "Mesoderm")
                      #"White_blood_cells", "Epithelium")

# 2. Run Stouffer's Method on the Geneset Dataframe
stouffer_gs_df <- res_df_gs %>%
  # Keep only static lineages (or whichever ones you intend to aggregate)
  #filter(!trajectory_refined %in% dynamic_lineages) %>%
  
  # Protect against p-values of exactly 0 (which cause infinite Z-scores in R)
  mutate(Spearman_pvalue = ifelse(Spearman_pvalue == 0, 1e-300, Spearman_pvalue)) %>%
  
  # Step A: Convert two-tailed p-values to directional Z-scores
  mutate(
    # qnorm converts p-value to Z-score. 
    # We multiply by the sign() of the Spearman R to keep the direction!
    Z_score = sign(Spearman) * qnorm(Spearman_pvalue / 2, lower.tail = FALSE)
  ) %>%
  
  # Step B: Group by geneset name to combine across trajectories
  group_by(gs_name) %>%
  summarize(
    mean_Spearman = mean(Spearman, na.rm = TRUE),
    median_Spearman = median(Spearman, na.rm = TRUE),
    k_trajectories = n(),                              # How many lineages was this geneset tested in?
    sum_Z = sum(Z_score, na.rm = TRUE),                # Add the Z-scores together
    n_positive_0.2 = sum(Spearman > 0.2, na.rm = TRUE), 
    n_positive_0.3 = sum(Spearman > 0.3, na.rm = TRUE), 
    n_positive_0.4 = sum(Spearman > 0.4, na.rm = TRUE), 
    n_negative = sum(Spearman < -0.3, na.rm = TRUE),     # Count negative votes
    .groups = "drop"
  ) %>%
  
  # Step C: Calculate Stouffer's Z and convert back to a Meta P-value
  mutate(
    Stouffer_Z = sum_Z / sqrt(k_trajectories),         # The core Stouffer formula
    Meta_Pvalue = 2 * pnorm(-abs(Stouffer_Z)),         # Convert combined Z back to a two-tailed p-value
    Meta_FDR = p.adjust(Meta_Pvalue, method = "BH")    # Apply the FDR penalty to the meta p-value
  ) %>%
  
  # Step D: Sort by the strongest consistent positive correlations
  arrange(desc(Stouffer_Z))

# View top consistently correlated genesets across your selected lineages
head(stouffer_gs_df, 5)
```

```r
# 1. Clean and deduplicate the whole genome list
stouffer_clean <- stouffer_df %>%
  # Remove any rows where the gene name is missing
  filter(!is.na(gene_short_name), gene_short_name != "") %>%
  # Convert to UPPER CASE first so "Myc" and "MYC" are treated as the same
  mutate(gene_name_upper = toupper(gene_short_name)) %>%
  # Group by the name and take the mean Z-score for duplicates
  group_by(gene_name_upper) %>%
  summarize(Stouffer_Z = mean(Stouffer_Z, na.rm = TRUE), .groups = "drop")

# 2. Create the unique ranks vector
ranks <- stouffer_clean$Stouffer_Z
names(ranks) <- stouffer_clean$gene_name_upper

# 3. Center the ranks and add tiny noise to break ties
# (This ensures the background is centered at zero)
ranks_final <- ranks - mean(ranks, na.rm = TRUE)
ranks_final <- ranks_final + rnorm(length(ranks_final), 0, 1e-9)
ranks_final <- sort(ranks_final, decreasing = TRUE)

# 4. Check the intersection again to be safe
# (Should be thousands now if using the whole genome)
print(paste("Unique genes for GSEA:", length(ranks_final)))
print(paste("Matching genes in Hallmark:", length(intersect(names(ranks_final), unlist(m_list_upper)))))
```

```r
rm_data <- metadata_fit %>%
    mutate(trajectory_refined = case_match(trajectory_refined,
                                           "White_blood_cells" ~ "Myeloid",
                                           "Muscle_cells" ~ "Myogenic",
                                           "Epithelium" ~ "Epithelial",
                                           "Fibroblasts_Fap+" ~ "Fibroblasts",
                                           "Hepatocytes" ~ "Hepatic",
                                           "Definitive_erythroid" ~ "Erythroid",
                                           "Upper_layer_neurons" ~ "Upper layer\nneurons",
                                           "GABAergic" ~ "GABAergic\nneurons",
                                           "Glutamatergic" ~ "Glutamatergic\nneurons",
                                           .default = trajectory_refined)) %>%
    # Ensure Embryo is a factor for rmcorr
    mutate(Embryo = factor(Embryo))

rm_ranked_results <- rm_data %>%
    # Remove rows with NA scores
    filter(!is.na(avg_score), !is.na(negative_genes)) %>%
    group_by(trajectory_refined) %>%
    group_modify(~ {
        # A. Filter for embryos that have AT LEAST 3 cells
        valid_embryos <- .x %>%
            count(Embryo) %>%
            filter(n >= 3) %>%
            pull(Embryo)
        
        df_sub <- .x %>% filter(Embryo %in% valid_embryos)
        
        # B. Check if we have at least 2 valid embryos left
        if (length(unique(df_sub$Embryo)) < 2) {
            return(data.frame(r_rm = NA, p_val = NA, n_embryos = length(unique(df_sub$Embryo))))
        }
        
        # C. Run Ranked rmcorr
        tryCatch({
            # Ranking within the trajectory group
            res <- rmcorr(participant = Embryo, 
                          measure1 = rank(avg_score), 
                          measure2 = rank(negative_genes), 
                          dataset = df_sub)
            
            data.frame(
                r_rm = round(res$r, 3),
                p_val = res$p,
                n_embryos = length(unique(df_sub$Embryo))
            )
        }, error = function(e) {
            data.frame(r_rm = NA, p_val = NA, n_embryos = length(unique(df_sub$Embryo)))
        })
    }) %>%
    ungroup()

rm_ranked_results <- rm_ranked_results %>%
    mutate(label = ifelse(is.na(r_rm), 
                          'Insuff. Data', # Valid expression for parse=TRUE
                          paste0("ρ = ", r_rm)))
        
# 3. View the results
print(rm_ranked_results)
```

```r
dynamic_lineages <- c("GABAergic", "Glutamatergic", "Mesoderm")
                      #"White_blood_cells", "Epithelium")

# 2. Run Stouffer's Method
stouffer_df <- res_df %>%
  # Keep only dynamic lineages
  filter(!trajectory_refined %in% dynamic_lineages) %>%
  
  # Protect against p-values of exactly 0 (which cause infinite Z-scores in R)
  mutate(Spearman_pvalue = ifelse(Spearman_pvalue == 0, 1e-300, Spearman_pvalue)) %>%
  
  # Step A: Convert two-tailed p-values to directional Z-scores
  mutate(
    # qnorm converts p-value to Z-score. 
    # We multiply by the sign() of the Spearman R to keep the direction!
    Z_score = sign(Spearman) * qnorm(Spearman_pvalue / 2, lower.tail = FALSE)
  ) %>%
  
  # Step B: Group by gene to combine across trajectories
  group_by(gene_ID, gene_short_name) %>%
  summarize(
    mean_Spearman = mean(Spearman, na.rm = TRUE),
    median_Spearman = median(Spearman, na.rm = TRUE),
    k_trajectories = n(),                             # How many lineages was this gene tested in?
    sum_Z = sum(Z_score, na.rm = TRUE),               # Add the Z-scores together
    n_positive_0.2 = sum(Spearman > 0.2, na.rm = TRUE), 
    n_positive_0.3 = sum(Spearman > 0.3, na.rm = TRUE), 
    n_positive_0.4 = sum(Spearman > 0.4, na.rm = TRUE), 
    n_negative = sum(Spearman < -0.3, na.rm = TRUE),     # Count negative votes
    Spearman = first(global_cor),
    .groups = "drop"
  ) %>%
  
  # Step C: Calculate Stouffer's Z and convert back to a Meta P-value
  mutate(
    Stouffer_Z = sum_Z / sqrt(k_trajectories),        # The core Stouffer formula
    Meta_Pvalue = 2 * pnorm(-abs(Stouffer_Z)),        # Convert combined Z back to a two-tailed p-value
    Meta_FDR = p.adjust(Meta_Pvalue, method = "BH")   # Finally, apply the FDR penalty to the meta p-value!
  ) %>%
  
  # Step D: Sort by the strongest consistent positive correlations
  arrange(desc(Stouffer_Z))
```

```r
stouffer_Z_5_median_0.7_0.4_n_6 = stouffer_df %>%
    filter(Stouffer_Z > 5 & Meta_FDR < 0.01 & n_negative < 2 & n_positive_0.3 > 5) %>%
    filter(median_Spearman > 0.4) %>%
    filter(Spearman > 0.7)

stouffer_Z_5_median_0.7_0.4_n_6
```

```r
set_name = "Z5_0.7_0.4_n_6"
# 1. PREPARE DATA
# Use the specific column created for this set
score_col <- paste0("gene_score_", set_name)

metadata_fit = as.data.frame(colData(pb_cds_nuc_frac_all))

# Ensure our x-axis variable is normalized for the model
metadata_fit$norm_x = normalize(metadata_fit[[score_col]])

# Squeeze the response variable (avg_score) to fit strictly within (0, 1)
n <- nrow(metadata_fit)
metadata_fit$fraction_squeezed <- (normalize(metadata_fit$avg_score) * (n - 1) + 0.5) / n

# 2. BUILD THE MIXED-EFFECTS MODEL
# Random effect (1 | Embryo) accounts for batch variation across replicates
model_glmm <- glmmTMB(
  fraction_squeezed ~ norm_x + (1 | Embryo), 
  data = metadata_fit, 
  family = beta_family(link = "cloglog")
)

# 3. PREDICT RESPONSE
# Generate predictions for the fixed effect across the 0-1 range
pred_universal <- predict_response(
    model_glmm, 
    terms = "norm_x [0:1, by=0.01]",
    type = "fixed"
)
```

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res=300)

ggplot() +
  # Points colored by Embryo to visualize the random effect spread
  geom_point(data = metadata_fit, 
             aes(x = norm_x, y = fraction_squeezed), 
             alpha = 0.8, size = 0.7) +

  # Fixed Effect Regression Line
  geom_line(data = pred_universal, 
            aes(x = x, y = predicted), 
            color = "red", linewidth = 0.8) +

  # 95% CI Ribbon
  geom_ribbon(data = pred_universal, 
              aes(x = x, ymin = conf.low, ymax = conf.high), 
              fill = "red", alpha = 0.15) +

  labs(x = "Normalized geneset expression", 
       y = "Nucleolar fraction score",
       color = "Embryo") +
  monocle3:::monocle_theme_opts() +
  theme(legend.position = "right", aspect.ratio = 1) + 
    theme(
        strip.text.x = element_text(size = 13), 
        strip.text.y = element_text(size = 8.5, angle = 270),
        strip.background = element_blank(),
        
        axis.text = element_text(color="black", size = 15),
        axis.ticks = element_line(color = "black"),
        axis.title = element_text(color = "black", size = 15)
    )
```
