```r
cds_VCS_pseudotime = readRDS("cds_VCS_with_cell_type_labels_E15-15.5_1e-4_subclustered_scores_by_embryo_30_all_genes_pseudotime.RDS")
```

## Astrocytes

```r
cds_VCS_sub = cds_VCS_pseudotime[, cds_VCS_pseudotime$trajectory_label == "Astrocytes"]
```

```r
options(repr.plot.width=8, repr.plot.height=5, repr.plot.res=300)
plot_cells(cds_VCS_sub, genes = c("Pla2g7", "Sox9", "Slc6a11"),
           x = 1, y = 2, scale_to_range = F,
           min_expr = 0.1,
           reduction_method = "UMAP", 
           color_cells_by = "nn_majority_cell_type_umap", 
           group_label_size = 3,
           label_cell_groups = F,
           cell_size = 0.4, alpha = 0.8) + 
        guides(color = guide_colorbar(title = "log10\nexpression")) + 
        theme(aspect.ratio = 1) +
        theme(strip.text = element_text(size = 12))

options(repr.plot.width=4, repr.plot.height=3, repr.plot.res=300)
plot_cells(cds_VCS_sub, x = 1, y = 2, show_trajectory_graph = F,
           color_cells_by = "Pseudotime")  + 
        guides(color = guide_colorbar(title = "Pseudotime")) + 
        theme(aspect.ratio = 1) +
        theme(strip.text = element_text(size = 12))
```

```r
umap_coordinates = reducedDims(cds_VCS_sub)[['UMAP']]
cds_VCS_sub$UMAP1 = umap_coordinates[, 1]
cds_VCS_sub$UMAP2 = umap_coordinates[, 2]
cds_VCS_sub$UMAP3 = umap_coordinates[, 3]
```

```r
metadata_sub = as.data.frame(colData(cds_VCS_sub))
```

```r
plot_df <- metadata_sub %>%
  filter(!Marker %in% c("Nuclear", "SiRDNA"))

# 2. Filter and Rename Labels
plot_df <- metadata_sub %>%
  filter(!Marker %in% c("Nuclear", "SiRDNA")) %>%
  mutate(
    # Change labels as requested
    Marker = case_when(
      Marker == "Nucleolin" ~ "Nucleolar",
      Marker == "SC35"      ~ "Speckle",
      TRUE ~ Marker
    ),
    Phenotype = case_when(
      Phenotype == "Ratio" ~ "Fraction",
      TRUE ~ Phenotype
    )
  )

# 3. Create the plotting function with local scales and requested aesthetics
make_umap_plot <- function(sub_df) {
        m_mid <- median(sub_df$avg_score, na.rm = TRUE)
        m_min <- quantile(sub_df$avg_score, 0.15, na.rm = TRUE)
        m_max <- quantile(sub_df$avg_score, 0.85, na.rm = TRUE)

        dist <- max(abs(m_mid - m_min), abs(m_max - m_mid))

        # Set symmetric limits
        m_min <- m_mid - dist
        m_max <- m_mid + dist

        # Format the title to look like a centered strip
        title_label <- paste0(unique(sub_df$Marker), " ", unique(sub_df$Phenotype))

        ggplot(sub_df, aes(x = UMAP1, y = UMAP2, color = avg_score)) +
        geom_point(size = 0.5, alpha = 0.8) +
        scale_colour_gradient2(
          name = "Score", 
          low = muted("blue"), 
          mid = "grey90", 
          high = muted("red"), 
          midpoint = m_mid,
          limits = c(m_min, m_max), 
          oob = scales::squish
        ) +
        labs(title = title_label, x = "UMAP 1", y = "UMAP 2") +
        #monocle3:::monocle_theme_opts() +
        theme(
            aspect.ratio = 1,
            panel.background = element_rect(fill = "white", color = NA),
            plot.background = element_rect(fill = "white", color = NA),
            panel.grid.major = element_blank(),
            panel.grid.minor = element_blank(),
            axis.line = element_blank(),
            axis.ticks = element_blank(),
            axis.text = element_blank(),
            axis.title = element_blank(),
            # Bold and centered title
            # Center and bold the title
            # Center the title (acting as strip text)
            plot.title = element_text(size = 15, hjust = 0.5),
            # Legend aesthetics
            legend.key.height = unit(0.6, "cm"),
            legend.key.width = unit(0.5, "cm"),
            legend.text = element_text(size = 12),
            legend.title = element_text(size = 14)
        )
}

# 4. Split data by new labels, generate plots, and assemble
plot_list <- plot_df %>%
  group_by(Marker, Phenotype) %>%
  group_split() %>%
  purrr::map(make_umap_plot)


options(repr.plot.width=16, repr.plot.height=7, repr.plot.res=300)
wrap_plots(plot_list, ncol = 4)
```
